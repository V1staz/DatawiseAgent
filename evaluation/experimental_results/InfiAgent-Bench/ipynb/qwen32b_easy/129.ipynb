[STEP GOAL]: Load the titanic.csv dataset and calculate the mean and standard deviation of the fare column.


````python
# Load the titanic.csv dataset
titanic_df = pd.read_csv('./input/titanic.csv')

# Display the first few rows of the dataframe to understand its structure
titanic_df.head()
````

````python
# Calculate the mean of the 'Fare' column
mean_fare = titanic_df['Fare'].mean()

# Calculate the population standard deviation of the 'Fare' column
std_deviation_fare = titanic_df['Fare'].std(ddof=0)  # ddof=0 for population standard deviation

mean_fare, std_deviation_fare
````

The mean fare paid by the passengers is approximately $32.20, and the population standard deviation of the fare is approximately $49.67. These statistics provide insight into the central tendency and dispersion of the fare paid by passengers on the Titanic.


[STEP GOAL]: Manually calculate the mean and population standard deviation of the fare column using the provided formulas.


````python
# Manually calculating the mean
total_fare = titanic_df['Fare'].sum()
count_fare = titanic_df['Fare'].count()
mean_fare_manual = total_fare / count_fare

# Manually calculating the population standard deviation
mean_fare = mean_fare_manual
variance_fare = ((titanic_df['Fare'] - mean_fare) ** 2).sum() / count_fare
std_deviation_fare_manual = variance_fare ** 0.5

mean_fare_manual, std_deviation_fare_manual
````

The manually calculated mean fare paid by the passengers is approximately $32.20, and the population standard deviation of the fare is approximately $49.67. This confirms our earlier calculations using the built-in functions in pandas.
