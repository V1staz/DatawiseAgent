[STEP GOAL]: Load the `estimated_numbers.csv` dataset, inspect its contents, and ensure the 'No. of deaths' column is properly formatted as numeric data.


````python
# Load the dataset
df = pd.read_csv('./input/estimated_numbers.csv')

# Display the first few rows of the DataFrame
df.head()
````

[STEP GOAL]: Convert the 'No. of deaths' column to numeric data type and handle any conversion errors.


````python
# Attempt to convert the 'No. of deaths' column to numeric values, coercing errors to NaN
df['No. of deaths'] = pd.to_numeric(df['No. of deaths'], errors='coerce')

# Fill NaN values with 0 (or drop these rows if necessary)
df['No. of deaths'].fillna(0, inplace=True)

# Verify the changes by displaying the DataFrame again
df.head()
````

````python
# Update the DataFrame using a safer method to avoid the FutureWarning
df['No. of deaths'] = df['No. of deaths'].fillna(0)

# Verify the changes by displaying the DataFrame again
df.head()
````

The conversion of the 'No. of deaths' column to numeric data type is now complete, and any non-numeric values have been coerced to NaN. These NaN values were subsequently filled with 0s to ensure no issues arise when performing calculations.


[STEP GOAL]: Calculate the maximum value in the 'No. of deaths' column and identify the corresponding country and year.


````python
# Find the row with the maximum value in the 'No. of deaths' column
max_deaths_row = df.loc[df['No. of deaths'].idxmax()]

# Extract the country, year, and max deaths value
country_with_max_deaths = max_deaths_row['Country']
year_of_max_deaths = max_deaths_row['Year']
max_deaths_value = max_deaths_row['No. of deaths']

(country_with_max_deaths, year_of_max_deaths, max_deaths_value)
````
