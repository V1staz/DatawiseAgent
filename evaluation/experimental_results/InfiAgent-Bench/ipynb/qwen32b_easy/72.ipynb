[STEP GOAL]: Perform the Shapiro-Wilk test on the 'Close' column to check for normal distribution.


````python
from scipy.stats import shapiro

# Extract the 'Close' column values
close_prices = df_microsoft['Close'].dropna()

# Perform the Shapiro-Wilk test
stat, p_value = shapiro(close_prices)

# Display the test statistic and p-value
(stat, p_value)
````

[STEP GOAL]: Interpret the results of the Shapiro-Wilk test and determine if the 'Close' column follows a normal distribution based on the p-value threshold of 0.05.


````python
# Define the significance level
alpha = 0.05

# Compare the p-value to the significance level to determine normality
if p_value > alpha:
    print("The 'Close' column appears to be normally distributed (p-value = {:.4f})".format(p_value))
else:
    print("The 'Close' column does not appear to be normally distributed (p-value = {:.4f})".format(p_value))
````

The Shapiro-Wilk test on the 'Close' column of the Microsoft stock prices has been conducted. With a p-value of approximately \( 1.41 \times 10^{-10} \), which is significantly less than the significance level of 0.05, we conclude that the 'Close' column does not adhere to a normal distribution.


The Shapiro-Wilk test was performed on the 'Close' column from the `microsoft.csv` dataset to check for normal distribution adherence. The test resulted in a p-value of approximately \( 1.41 \times 10^{-10} \), which is much lower than the significance level of 0.05. Based on this result, we can confidently state that the 'Close' column does not follow a normal distribution.
