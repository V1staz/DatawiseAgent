````python
import pandas as pd

# Load the CSV file into a DataFrame
file_path = "./input/20170413_000000_group_statistics.csv"
df = pd.read_csv(file_path)

# Correct the column name based on the printed column names
correct_column_name = 'num. calls abandoned '

# Remove any rows with null or NaN values in the correct column name
df_cleaned = df.dropna(subset=[correct_column_name])

# Calculate the total number of abandoned calls
total_abandoned_calls = df_cleaned[correct_column_name].sum()
print(f'Total number of abandoned calls: {total_abandoned_calls}')
````
