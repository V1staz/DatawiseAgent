````python
# Remove rows with missing data in the 'No. of cases' column
df_americas_cleaned = df_americas.dropna(subset=['No. of cases'])

# Extract the median value from the 'No. of cases' column
df_americas_cleaned['No. of cases'] = df_americas_cleaned['No. of cases'].str.extract('(\d+)', expand=False).astype(int)

# Group by 'Country' and calculate the mean of 'No. of cases'
average_cases_per_country = df_americas_cleaned.groupby('Country')['No. of cases'].mean().reset_index()

# Find the country with the highest average number of cases
country_with_highest_average_cases = average_cases_per_country.loc[average_cases_per_country['No. of cases'].idxmax()]

# Display the result
print(country_with_highest_average_cases)
````
