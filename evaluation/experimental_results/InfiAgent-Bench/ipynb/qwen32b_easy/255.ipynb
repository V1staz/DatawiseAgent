[STEP GOAL]: Load the CSV file and calculate the mean and standard deviation of the 'gdpPercap_2007' column. Round the results to 2 decimal places.


````python
import pandas as pd

# Load the dataset
df = pd.read_csv('./input/gapminder_gdp_asia.csv')

# Show the first few rows of the dataframe to understand its structure
df.head()
````

````python
# Calculate the mean and standard deviation of the 'gdpPercap_2007' column and round to 2 decimal places
mean_2007 = df['gdpPercap_2007'].mean().round(2)
std_dev_2007 = df['gdpPercap_2007'].std().round(2)

mean_2007, std_dev_2007
````

The mean and standard deviation of the 'gdpPercap_2007' column in the dataset have been calculated. The mean GDP per capita in 2007 across all countries in the dataset is **$12,473.03**, and the standard deviation is **$14,154.94**. Both values are rounded to two decimal places.
