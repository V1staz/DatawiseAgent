## Prompt

```text
Question: Calculate the mean and standard deviation of the "Income" column in the Credit.csv file.
Calculate the mean and standard deviation using built-in functions in Python's pandas library. Round the outputs to two decimal places.
```

## Expected Format

```text
@mean_income[mean_value], @std_dev_income[std_dev_value] where "mean_value" and "std_dev_value" are the calculated mean and standard deviation of the "Income" column, respectively. Both should be rounded to two decimal places.
```


[STEP GOAL]: Load the Credit.csv file into a pandas DataFrame and calculate the mean and standard deviation of the "Income" column.


````python
import pandas as pd

# Load the Credit.csv file into a DataFrame
credit_df = pd.read_csv('./input/Credit.csv')

# Calculate mean and standard deviation of the "Income" column
mean_income = credit_df['Income'].mean()
std_income = credit_df['Income'].std()

# Round the outputs to two decimal places
mean_income_rounded = round(mean_income, 2)
std_income_rounded = round(std_income, 2)

mean_income_rounded, std_income_rounded
````

The mean income is 45.22 and the standard deviation of the income is 35.24, rounded to two decimal places.


The mean and standard deviation of the "Income" column in the Credit.csv file have been calculated and rounded to two decimal places. The mean income is 45.22, and the standard deviation of the income is 35.24.
