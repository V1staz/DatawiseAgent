````python
import pandas as pd
from scipy.stats import normaltest

# Load the dataset
df = pd.read_csv('./input/gapminder_cleaned.csv')

# Perform normality test on the "gdppercap" column
statistic, p_value = normaltest(df['gdppercap'])

p_value
````

The p-value obtained from the normality test is 0.0, which indicates that we reject the null hypothesis that the distribution of GDP per capita ("gdppercap") adheres to a normal distribution at the 0.05 significance level.


````python
# Interpretation of the p-value
is_normal_distribution = p_value > 0.05
is_normal_distribution
````

The normality test on the "gdppercap" column has been completed. The p-value is 0.0, indicating that the distribution of GDP per capita does not adhere to a normal distribution at the 0.05 significance level. This conclusion is based on the fact that the p-value is less than 0.05, leading us to reject the null hypothesis of normality.
