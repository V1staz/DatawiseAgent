[STEP GOAL]: Load the CSV file `test_x.csv` into a DataFrame and check its contents to understand the structure of the dataset.


````python
# Load the dataset into a DataFrame
df = pd.read_csv('./input/test_x.csv')

# Display the first few rows of the DataFrame
df.head()
````

[STEP GOAL]: Add a new feature "FamilySize" based on the provided formula.
Create the "FamilySize" column by summing the "IsAlone" column with a base value of 2 (one for siblings/spouses and one for parents/children).


````python
# Calculate the FamilySize based on the provided instruction
df['FamilySize'] = df['IsAlone'] + 2

# Display the updated DataFrame to confirm the addition of the FamilySize column
df.head()
````

[STEP GOAL]: Compute the average FamilySize and round the result to one decimal place.


````python
# Calculate and print the average FamilySize
average_family_size = round(df['FamilySize'].mean(), 1)
average_family_size
````
