[STEP GOAL]: Load the CSV file 'Current_Logan.csv' into a pandas DataFrame and calculate the percentage of missing values in the "Unnamed: 8" column.


````python
# Import pandas library
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('./input/Current_Logan.csv')

# Display the first few rows of the DataFrame to understand its structure
df.head()
````

[STEP GOAL]: Calculate the percentage of missing values in the "Unnamed: 8" column.


````python
# Calculate the percentage of missing values in the "Unnamed: 8" column
missing_percentage = df['Unnamed: 8'].isna().mean() * 100

missing_percentage
````
