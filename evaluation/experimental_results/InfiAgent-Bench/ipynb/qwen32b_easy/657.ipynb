## Prompt

```text
Question: Calculate the mean, median, and standard deviation of the 'Close' column.
Using pandas library in Python, ignore the missing values, and round the results to 2 decimal places.
```

## Expected Format

```text
@mean_close[mean], @median_close[median], @std_close[std_deviation] where "mean", "median", and "std_deviation" are decimal numbers representing the mean, median, and standard deviation of the 'Close' column, respectively, rounded to two decimal places.
```


[STEP GOAL]: Load the CSV file into a DataFrame and calculate the mean, median, and standard deviation of the 'Close' column while ignoring any missing values. Round the results to 2 decimal places.


````python
import pandas as pd

# Load the dataset
file_path = './input/YAHOO-BTC_USD_D.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataframe to understand its structure
data.head()
````

````python
# Calculate mean, median, and standard deviation of the 'Close' column
mean_close = data['Close'].mean(skipna=True)
median_close = data['Close'].median(skipna=True)
std_dev_close = data['Close'].std(skipna=True)

# Round the results to 2 decimal places
mean_close_rounded = round(mean_close, 2)
median_close_rounded = round(median_close, 2)
std_dev_close_rounded = round(std_dev_close, 2)

mean_close_rounded, median_close_rounded, std_dev_close_rounded
````

The mean, median, and standard deviation of the 'Close' column in the dataset are as follows, rounded to 2 decimal places:
- Mean: 4349.27
- Median: 3599.77
- Standard Deviation: 4113.51


The mean, median, and standard deviation of the 'Close' column in the dataset 'YAHOO-BTC_USD_D.csv' were successfully calculated and rounded to two decimal places. Here are the results:

- Mean: 4349.27
- Median: 3599.77
- Standard Deviation: 4113.51
