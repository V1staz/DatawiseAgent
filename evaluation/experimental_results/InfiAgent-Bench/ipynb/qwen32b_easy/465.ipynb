[STEP GOAL]: Load the dataset 'arrest_expungibility.csv' and calculate the skewness of the 'Age' column to determine if the distribution is approximately symmetric or skewed.


````python
import pandas as pd
from scipy.stats import skew

# Load the dataset
file_path = './input/arrest_expungibility.csv'
arrest_data = pd.read_csv(file_path)

# Display the first few rows of the dataframe to verify it has loaded correctly
arrest_data.head()
````

````python
# Calculate skewness of the 'Age' column
age_skewness = skew(arrest_data['Age'])
age_skewness
````
