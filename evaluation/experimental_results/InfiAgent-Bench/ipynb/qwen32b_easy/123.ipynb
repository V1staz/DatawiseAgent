[STEP GOAL]: Load the dataset and check for any missing values in the 'daily_vaccinations_per_million' column.


````python
# Load the dataset
df_vaccinations = pd.read_csv('./input/country_vaccinations.csv')

# Check for missing values in 'daily_vaccinations_per_million' column
missing_values = df_vaccinations['daily_vaccinations_per_million'].isnull().sum()
print(f"Missing values in 'daily_vaccinations_per_million': {missing_values}")
````

[STEP GOAL]: Filter out rows with missing values in the 'daily_vaccinations_per_million' column and calculate the average number of daily vaccinations per million for each country.


````python
# Filter out rows with missing values in 'daily_vaccinations_per_million'
filtered_df = df_vaccinations.dropna(subset=['daily_vaccinations_per_million'])

# Group by 'country' and calculate the mean of 'daily_vaccinations_per_million'
average_vaccinations = filtered_df.groupby('country')['daily_vaccinations_per_million'].mean().reset_index()

# Rename the columns for clarity
average_vaccinations.columns = ['country', 'avg_daily_vaccinations_per_million']

# Display the first few rows of the resulting DataFrame
average_vaccinations.head()
````

````python
# Find the country with the highest average number of daily vaccinations per million
highest_avg_country = average_vaccinations.loc[average_vaccinations['avg_daily_vaccinations_per_million'].idxmax()]
highest_avg_country
````

The country with the highest average number of daily vaccinations per million people is Gibraltar, with an average of approximately 21,999.73 vaccinations per million.
