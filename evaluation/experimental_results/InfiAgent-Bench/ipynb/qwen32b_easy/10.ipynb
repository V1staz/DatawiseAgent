[STEP GOAL]: Use the Shapiro-Wilk test from the scipy.stats module to check if the 'Total Traded Quantity' column follows a normal distribution.


````python
from scipy.stats import shapiro

# Extract the 'Total Traded Quantity' column
total_traded_quantity = df['Total Traded Quantity']

# Perform the Shapiro-Wilk test
stat, p_value = shapiro(total_traded_quantity)

# Print the test statistic and p-value
print(f'Statistic: {stat}, P-value: {p_value}')

# Determine if the data follows a normal distribution based on the p-value and alpha level
alpha = 0.05
if p_value > alpha:
    print("The data may follow a normal distribution (Failed to reject H0)")
else:
    print("The data does not follow a normal distribution (Reject H0)")
````

The Shapiro-Wilk test has been conducted on the 'Total Traded Quantity' column of the dataset. The p-value obtained is extremely low (approximately \(6.58 \times 10^{-27}\)), which is much lower than our chosen significance level (\(\alpha = 0.05\)). Therefore, we reject the null hypothesis that the data comes from a normally distributed population. This indicates that the 'Total Traded Quantity' does not adhere to a normal distribution.


The Shapiro-Wilk test has been applied to the 'Total Traded Quantity' column from the GODREJIND.csv dataset to determine if it follows a normal distribution. Based on the test statistic and p-value obtained, the data clearly does not adhere to a normal distribution (p-value ≈ \(6.58 \times 10^{-27}\), far below the 0.05 significance level). This conclusion means that we reject the null hypothesis, which posits that the data was drawn from a normal distribution. Therefore, it can be stated with confidence that the 'Total Traded Quantity' does not follow a normal distribution.
