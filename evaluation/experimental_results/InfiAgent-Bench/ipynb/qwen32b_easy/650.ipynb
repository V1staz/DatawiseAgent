````python
# Load the dataset
file_path = './input/DES=+2006261.csv'
dataset = pd.read_csv(file_path)

# Extract the X and Y coordinates columns
x_coords = dataset['X']
y_coords = dataset['Y']

# Calculate the Pearson correlation coefficient
correlation_coefficient = x_coords.corr(y_coords, method='pearson')

# Round the correlation coefficient to three decimal places
correlation_coefficient_rounded = round(correlation_coefficient, 3)

# Check if the correlation is negligible
if abs(correlation_coefficient_rounded) < 0.05:
    correlation_coefficient_rounded = 0

correlation_coefficient_rounded
````

The Pearson correlation coefficient between the X-coordinate and Y-coordinate columns has been calculated, rounded to three decimal places, and adjusted for negligible correlation values. The resulting correlation coefficient is **-0.868**, indicating a strong negative correlation between the two coordinates.
