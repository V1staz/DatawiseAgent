````python
# Load the dataset
df = pd.read_csv('./input/YAHOO-BTC_USD_D.csv')

# Calculate the Pearson correlation coefficient between the 'High' and 'Low' columns, ignoring missing values
correlation = df['High'].corr(df['Low'], method='pearson')

# Round the result to 2 decimal places
correlation_rounded = round(correlation, 2)
print(f"The Pearson correlation coefficient between 'High' and 'Low' is: {correlation_rounded}")
````

The Pearson correlation coefficient between the 'High' and 'Low' columns of the dataset has been calculated and rounded to 2 decimal places. The result is 1.0, indicating a perfect positive correlation between these two columns.
