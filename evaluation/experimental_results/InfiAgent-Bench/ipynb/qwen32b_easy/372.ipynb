[STEP GOAL]: Load the dataset and calculate the mean and median for the "Trips over the past 24-hours (midnight to 11:59pm)" column, rounding the results to two decimal places.


````python
import pandas as pd

# Load the dataset
file_path = "./input/2014_q4.csv"
data = pd.read_csv(file_path)

# Display the first few rows of the dataframe
data.head()
````

````python
# Calculate the mean and median of the "Trips over the past 24-hours (midnight to 11:59pm)" column
mean_trips = round(data["Trips over the past 24-hours (midnight to 11:59pm)"].mean(), 2)
median_trips = round(data["Trips over the past 24-hours (midnight to 11:59pm)"].median(), 2)

mean_trips, median_trips
````

The mean number of trips over the past 24 hours is **21144.08** and the median is **19711.00**, both values rounded to two decimal places.


The mean and median of the "Trips over the past 24-hours (midnight to 11:59pm)" column in the 2014_q4.csv dataset have been calculated successfully. Both values are rounded to two decimal places:

- Mean: 21144.08
- Median: 19711.00

This completes the requested analysis.
