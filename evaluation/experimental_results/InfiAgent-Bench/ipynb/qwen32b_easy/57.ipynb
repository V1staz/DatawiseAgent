````python
# Load the dataset
data_path = './input/estimated_numbers.csv'
df = pd.read_csv(data_path)

# Convert 'No. of cases' and 'No. of deaths' columns to integer type by extracting the median value
df['No. of cases'] = df['No. of cases'].str.extract(r'(\d+)').astype('int64')
df['No. of deaths'] = df['No. of deaths'].str.extract(r'(\d+)').astype('int64')

# Calculate the Pearson correlation coefficient
correlation_coefficient = df['No. of cases'].corr(df['No. of deaths'])

correlation_coefficient
````

The dataset has been loaded successfully. The 'No. of cases' and 'No. of deaths' columns have been converted to integer type by extracting the median values from the ranges provided. We calculated the Pearson correlation coefficient, which is approximately 0.9697. This indicates a strong positive correlation between the number of cases and the number of deaths recorded in the dataset.
