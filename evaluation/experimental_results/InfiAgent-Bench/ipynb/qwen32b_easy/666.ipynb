```text
Question: Calculate the mean and standard deviation of the MedianHouseValue column in the provided dataset.
Calculate the mean and standard deviation to four decimal places using built-in Python statistical functions.
```

## Expected Format

[STEP GOAL]: Load the dataset from `my_test_01.csv` and calculate the mean and standard deviation of the MedianHouseValue column to four decimal places.


````python
import pandas as pd

# Load the dataset
file_path = './input/my_test_01.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataframe to confirm it loaded correctly
data.head()
````

````python
# Calculate mean and standard deviation of MedianHouseValue column
mean_value = data['MedianHouseValue'].mean()
std_dev_value = data['MedianHouseValue'].std()

(mean_value, std_dev_value)
````

````python
# Round the mean and standard deviation to four decimal places
mean_value_rounded = round(mean_value, 4)
std_dev_value_rounded = round(std_dev_value, 4)

(mean_value_rounded, std_dev_value_rounded)
````

The mean and standard deviation of the MedianHouseValue column in the dataset have been calculated and rounded to four decimal places. The mean value is 2.1226 and the standard deviation is 1.2210.
