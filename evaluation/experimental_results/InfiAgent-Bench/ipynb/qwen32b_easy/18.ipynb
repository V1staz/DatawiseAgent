## Prompt

```text
Question: Calculate the mean and standard deviation of the "Mar.2019" column.
Outliers are to be pruned via the interquartile range method before calculating the mean and standard deviation. Handle missing values by using listwise deletion method. Report the measures rounded to two decimal places.
```

## Expected Format

```text
@mean_mar_2019[mean] @sd_mar_2019[sd], where "mean" and "sd" are numbers, rounded to two decimal places.
```


[STEP GOAL]: Load the dataset and inspect its structure.


````python
# Import necessary library
import pandas as pd

# Load the dataset
file_path = './input/unemployement_industry.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the dataframe
df.head()
````

[STEP GOAL]: Perform listwise deletion on the dataset and calculate the interquartile range for the "Mar.2019" column to identify outliers.


````python
# Drop rows with any missing values
df_cleaned = df.dropna()

# Calculate the IQR for the Mar.2019 column
Q1 = df_cleaned['Mar.2019'].quantile(0.25)
Q3 = df_cleaned['Mar.2019'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out the outliers
df_filtered = df_cleaned[(df_cleaned['Mar.2019'] >= lower_bound) & (df_cleaned['Mar.2019'] <= upper_bound)]

# Display the filtered dataframe
df_filtered.head()
````

[STEP GOAL]: Calculate the mean and standard deviation of the 'Mar.2019' column after outlier removal and rounding the results to two decimal places.


````python
# Calculate mean and standard deviation of 'Mar.2019' without outliers
mean_mar2019 = df_filtered['Mar.2019'].mean()
std_mar2019 = df_filtered['Mar.2019'].std()

# Round the measures to two decimal places
mean_mar2019_rounded = round(mean_mar2019, 2)
std_mar2019_rounded = round(std_mar2019, 2)

mean_mar2019_rounded, std_mar2019_rounded
````

The mean and standard deviation of the 'Mar.2019' column after removing outliers and applying listwise deletion for missing values are 171.44 and 188.25, respectively. These values have been rounded to two decimal places.


The mean and standard deviation of the 'Mar.2019' column, after pruning outliers using the interquartile range method and handling missing values through listwise deletion, are calculated to be 171.44 and 188.25, respectively. These results have been rounded to two decimal places. This completes the fulfillment of the [USER INSTRUCTION].
