[STEP GOAL]: Load the Titanic dataset and calculate the Pearson correlation coefficient between the passenger class (Pclass) and the fare (Fare), excluding rows with null values.


````python
# Load the dataset
df_titanic = pd.read_csv('./input/titanic.csv')

# Display the first few rows of the DataFrame to understand its structure
df_titanic.head()
````

````python
# Calculate the Pearson correlation coefficient between 'Pclass' and 'Fare', excluding rows with null values
correlation_coefficient = df_titanic[['Pclass', 'Fare']].corr(method='pearson').iloc[0, 1]
correlation_coefficient_rounded = round(correlation_coefficient, 2)
correlation_coefficient_rounded
````

The Pearson correlation coefficient between passenger class (`Pclass`) and fare (`Fare`), excluding rows with null values, is **-0.55**. This indicates a moderate negative correlation between the passenger class and the fare.
