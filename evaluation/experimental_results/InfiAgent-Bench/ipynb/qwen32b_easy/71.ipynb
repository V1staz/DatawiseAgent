[STEP GOAL]: Load the 'microsoft.csv' dataset into a pandas DataFrame and calculate the mean and standard deviation of the 'Volume' column.


````python
import pandas as pd

# Load the dataset
df = pd.read_csv('./input/microsoft.csv')

# Display the first few rows of the dataframe to understand its structure
df.head()
````

````python
# Calculate the mean and standard deviation of the 'Volume' column
mean_volume = df['Volume'].mean()
std_volume = df['Volume'].std()

# Round the results to 2 decimal places
mean_volume_rounded = round(mean_volume, 2)
std_volume_rounded = round(std_volume, 2)

mean_volume_rounded, std_volume_rounded
````

The mean and standard deviation of the 'Volume' column from the 'microsoft.csv' dataset have been calculated successfully. The mean is approximately 22,607,406.19 and the standard deviation is approximately 8,254,791.71, both rounded to two decimal places.
