[STEP GOAL]: Perform the Shapiro-Wilk test for normality on the 'per_dem' column to determine if it follows a normal distribution. Set the significance level (alpha) at 0.05.


````python
from scipy.stats import shapiro

# Extract the 'per_dem' column for the Shapiro-Wilk test
dem_percentage = df_election['per_dem']

# Perform the Shapiro-Wilk test
stat, p_value = shapiro(dem_percentage)

# Define the significance level
alpha = 0.05

# Check the result against the significance level
if p_value < alpha:
    print(f"The data does not follow a normal distribution (p-value: {p_value:.4f})")
else:
    print(f"The data follows a normal distribution (p-value: {p_value:.4f})")
````

The Shapiro-Wilk test for normality has been performed on the 'per_dem' column of the election2016 dataset. The resulting p-value was less than the significance level of 0.05, indicating that we can reject the null hypothesis. Therefore, it can be concluded that the percentage of votes received by the Democratic party in a particular county is not normally distributed.
