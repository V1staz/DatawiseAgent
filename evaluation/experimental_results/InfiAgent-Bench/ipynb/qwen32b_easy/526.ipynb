Question: Is there a correlation between the passenger class and the fare paid?

[STEP GOAL]: Load the Titanic test dataset and calculate the Pearson correlation coefficient between 'Pclass' and 'Fare'.


````python
# Load the Titanic test dataset
titanic_df = pd.read_csv('./input/titanic_test.csv')

# Display the first few rows of the DataFrame to understand its structure
titanic_df.head()
````

[STEP GOAL]: Calculate the Pearson correlation coefficient between 'Pclass' and 'Fare', ignoring rows with missing values in these two columns. Round the result to two decimal places.


````python
# Calculate the Pearson correlation coefficient between 'Pclass' and 'Fare'
correlation_coefficient = titanic_df['Pclass'].corr(titanic_df['Fare'], method='pearson')
correlation_coefficient_rounded = round(correlation_coefficient, 2)

# Display the correlation coefficient rounded to two decimal places
correlation_coefficient_rounded
````

The Pearson correlation coefficient between 'Pclass' and 'Fare' is approximately **-0.58**, indicating a moderate negative linear relationship. This suggests that as the passenger class decreases (higher numeric value), the fare tends to decrease as well.


The Pearson correlation coefficient between 'Pclass' and 'Fare' has been calculated, showing a correlation coefficient of approximately **-0.58**. This indicates a moderate negative linear relationship, meaning that higher passenger classes ('Pclass') are generally associated with lower fares, as expected given that higher class numbers typically represent lower socio-economic status on the Titanic dataset.
