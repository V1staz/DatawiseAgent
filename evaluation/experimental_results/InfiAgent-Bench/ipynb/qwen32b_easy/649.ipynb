[STEP GOAL]: Load the CSV file and calculate the mean and standard deviation of the X-coordinate column, rounding to three decimal places.


````python
import pandas as pd

# Load the dataset
file_path = "./input/DES=+2006261.csv"
data = pd.read_csv(file_path)

# Display the first few rows to understand the structure
data.head()
````

````python
# Assuming the X-coordinate column is named 'X'
mean_x = round(data['X'].mean(), 3)
std_dev_x = round(data['X'].std(), 3)

print(f"Mean of X-coordinate: {mean_x}")
print(f"Standard Deviation of X-coordinate: {std_dev_x}")
````

The mean and standard deviation of the X-coordinate column have been successfully calculated and rounded to three decimal places. The results are as follows:
- Mean: -259162995.016
- Standard Deviation: 53529181.172


[STEP GOAL]: Verify if the built-in Python functions were used correctly to calculate the mean and standard deviation of the X-coordinate column and round these values to three decimal places.


````python
import statistics

# Using built-in Python functions to calculate the mean and standard deviation
mean_builtin = round(statistics.mean(data['X']), 3)
std_dev_builtin = round(statistics.stdev(data['X']), 3)

print(f"Built-in Mean of X-coordinate: {mean_builtin}")
print(f"Built-in Standard Deviation of X-coordinate: {std_dev_builtin}")
````

The built-in Python functions have been used correctly to calculate the mean and standard deviation of the X-coordinate column, and the values have been rounded to three decimal places. The results match those obtained using Pandas, confirming the calculations are accurate.
