[STEP GOAL]: Load the dataset and inspect its structure to understand the columns and the data it contains.


````python
import pandas as pd

# Load the dataset
file_path = './input/percent-bachelors-degrees-women-usa.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
data.head()
````

````python
# Check the structure of the dataset
data.info()
````

The dataset has been successfully loaded and inspected. It contains 42 entries, with each entry representing a year from 1970 to 2011. The dataset includes the percentage of women who earned bachelor's degrees in various fields.

Next, we will focus on the "Architecture" field and perform outlier detection using the Z-score method with a threshold of 3.


[STEP GOAL]: Perform outlier detection on the "Architecture" field using the Z-score method with a threshold of 3.


````python
import numpy as np

# Extract the "Architecture" column
architecture_data = data['Architecture']

# Calculate the Z-scores
z_scores = (architecture_data - architecture_data.mean()) / architecture_data.std()

# Identify outliers
outliers = (np.abs(z_scores) > 3)
outlier_years = data['Year'][outliers]

# Display the years with outliers
outlier_years
````

````python
# Remove outliers and calculate the mean and standard deviation for the remaining data
non_outlier_data = architecture_data[~outliers]
mean_percentage = non_outlier_data.mean().round(2)
std_deviation = non_outlier_data.std().round(2)

mean_percentage, std_deviation
````

[STEP GOAL]: Summarize the results and provide the final output, including the identified outlier years (if any), the mean percentage, and the standard deviation for the years without outliers.


The dataset did not contain any outliers in the "Architecture" field based on the Z-score method with a threshold of 3. Therefore, all years were included in the calculation of the mean percentage and standard deviation.

The mean percentage of women earning bachelor's degrees in Architecture from 1970 to 2011 is **33.69%**, and the standard deviation is **9.57%**.
