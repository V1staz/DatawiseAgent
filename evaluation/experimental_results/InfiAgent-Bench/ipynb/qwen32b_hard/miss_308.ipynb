## Prompt

```text
Question: Use feature engineering techniques to create a new variable "Title" by extracting the title from the Name column (e.g., "Mr.", "Mrs.", "Miss"). Only consider the following titles: 'Mr.', 'Mrs.', 'Miss.' and 'Master.' (titles followed by a dot). Then, calculate the average fare for each unique title to two decimal places.
Only the titles 'Mr.', 'Mrs.', 'Miss.' and 'Master.' should be considered. Titles that do not fall within these four categories should be eliminated.
```

## Expected Format

```text
@average_fare_Mr[value1], @average_fare_Mrs[value2], @average_fare_Miss[value3], @average_fare_Master[value4], where value1, value2, value3, and value4 represent the average fares for 'Mr.', 'Mrs.', 'Miss.', and 'Master.', respectively. All values should be rounded to two decimal places.
```


[STEP GOAL]: Load the Titanic dataset and inspect its initial structure to understand the columns and the first few rows of data.


````python
# Import pandas library
import pandas as pd

# Load the dataset
titanic_df = pd.read_csv('./input/titanic.csv')

# Display the first few rows of the dataframe
titanic_df.head()
````

[STEP GOAL]: Extract the "Title" from the "Name" column and filter out any titles that are not 'Mr.', 'Mrs.', 'Miss.', or 'Master.'


````python
# Define a function to extract the title from the name
def extract_title(name):
    import re
    # Define the list of valid titles
    valid_titles = ['Mr.', 'Mrs.', 'Miss.', 'Master.']
    # Use regular expression to find the title
    match = re.search(r'\b(' + '|'.join(valid_titles) + r')\b', name)
    # Return the matched title or None if no valid title is found
    return match.group(0) if match else None

# Apply the function to the 'Name' column to create the 'Title' column
titanic_df['Title'] = titanic_df['Name'].apply(extract_title)

# Drop rows where 'Title' is None (not in the valid titles)
titanic_df = titanic_df.dropna(subset=['Title'])

# Display the first few rows of the updated dataframe
titanic_df.head()
````

[STEP GOAL]: Calculate the average fare for each unique title and round the result to two decimal places.


````python
# Group the data by 'Title' and calculate the mean fare for each group
average_fare_by_title = titanic_df.groupby('Title')['Fare'].mean().round(2)

# Display the result
average_fare_by_title
````

The average fare for each unique title in the Titanic dataset, rounded to two decimal places, is as follows:

- **Mr.**: $34.01
- **Mrs.**: $44.73

Note: Only the titles 'Mr.', 'Mrs.', 'Miss.', and 'Master.' were considered, and titles that did not fall within these categories were eliminated.
