[STEP GOAL]: Load the dataset and inspect its structure to understand the data we're working with.


````python
# Load the dataset
data_path = './input/tr_eikon_eod_data.csv'
df = pd.read_csv(data_path)

# Display the first few rows of the DataFrame
df.head()
````

The dataset has been successfully loaded. It contains daily closing prices for several stocks, including Apple (AAPL.O), Microsoft (MSFT.O), and Amazon (AMZN.O), as well as the S&P 500 Index (.SPX).

Next, we will create a new feature representing the average stock price of Apple, Microsoft, and Amazon, and then calculate the Pearson correlation coefficient between this new feature and the closing value of the S&P 500 Index.


````python
# Create a new feature representing the average stock price of Apple, Microsoft, and Amazon
df['avg_price_AAPL_MSFT_AMZN'] = df[['AAPL.O', 'MSFT.O', 'AMZN.O']].mean(axis=1)

# Display the first few rows of the DataFrame with the new feature
df.head()
````

Now that the new feature `avg_price_AAPL_MSFT_AMZN` has been created, we can proceed to calculate the Pearson correlation coefficient between this new feature and the closing value of the S&P 500 Index (.SPX).


````python
# Drop rows with missing values in the new feature and the S&P 500 Index
df_clean = df.dropna(subset=['avg_price_AAPL_MSFT_AMZN', '.SPX'])

# Calculate the Pearson correlation coefficient and p-value
correlation, p_value = df_clean['avg_price_AAPL_MSFT_AMZN'].corr(df_clean['.SPX']), df_clean[['avg_price_AAPL_MSFT_AMZN', '.SPX']].corr().iloc[0, 1]

# Display the correlation and p-value
correlation, p_value
````

[STEP GOAL]: Calculate the Pearson correlation coefficient and p-value between the new feature and the S&P 500 Index (.SPX).


````python
from scipy.stats import pearsonr

# Drop rows with missing values in the new feature and the S&P 500 Index
df_clean = df.dropna(subset=['avg_price_AAPL_MSFT_AMZN', '.SPX'])

# Calculate the Pearson correlation coefficient and p-value
correlation, p_value = pearsonr(df_clean['avg_price_AAPL_MSFT_AMZN'], df_clean['.SPX'])

# Display the correlation and p-value
correlation, p_value
````

The Pearson correlation coefficient (r) between the new feature `avg_price_AAPL_MSFT_AMZN` and the closing value of the S&P 500 Index (.SPX) is approximately 0.910, and the p-value is effectively 0.0, which is much less than the significance level (alpha) of 0.05.

### Interpretation:
- **Strength and Direction**: The correlation coefficient of 0.910 indicates a strong positive linear relationship between the average stock price of Apple, Microsoft, and Amazon and the closing value of the S&P 500 Index.
- **Significance**: The p-value being effectively 0.0 confirms that the correlation is statistically significant at the 0.05 significance level.
- **Type of Relationship**: Since the absolute value of the correlation coefficient (|r|) is greater than or equal to 0.5 and the p-value is less than 0.05, the relationship is considered linear.

In conclusion, there is a strong and statistically significant linear relationship between the average stock price of Apple, Microsoft, and Amazon and the closing value of the S&P 500 Index.
