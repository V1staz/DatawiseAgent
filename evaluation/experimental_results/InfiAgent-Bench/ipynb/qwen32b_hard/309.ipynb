[STEP GOAL]: Load the Titanic dataset and inspect its contents to understand the structure and initial statistics of the 'Age' and 'Fare' columns.


````python
# Import necessary libraries
import pandas as pd

# Load the Titanic dataset
file_path = './input/titanic.csv'
titanic_df = pd.read_csv(file_path)

# Display the first few rows of the dataset
titanic_df.head()
````

````python
# Display summary statistics for the 'Age' and 'Fare' columns
titanic_df[['Age', 'Fare']].describe()
````

[STEP GOAL]: Perform distribution analysis on the 'Age' and 'Fare' variables, including calculating skewness and kurtosis, and counting the number of values within one standard deviation from the mean.


````python
import scipy.stats as stats

# Calculate skewness and kurtosis for 'Age' and 'Fare'
age_skewness = stats.skew(titanic_df['Age'].dropna())
age_kurtosis = stats.kurtosis(titanic_df['Age'].dropna())

fare_skewness = stats.skew(titanic_df['Fare'])
fare_kurtosis = stats.kurtosis(titanic_df['Fare'])

# Calculate mean and standard deviation for 'Age' and 'Fare'
age_mean = titanic_df['Age'].mean()
age_std = titanic_df['Age'].std()

fare_mean = titanic_df['Fare'].mean()
fare_std = titanic_df['Fare'].std()

# Count the number of values within one standard deviation from the mean for 'Age' and 'Fare'
age_within_1_std = ((titanic_df['Age'] >= age_mean - age_std) & (titanic_df['Age'] <= age_mean + age_std)).sum()
fare_within_1_std = ((titanic_df['Fare'] >= fare_mean - fare_std) & (titanic_df['Fare'] <= fare_mean + fare_std)).sum()

# Print the results
print(f"Age Skewness: {age_skewness:.4f}")
print(f"Age Kurtosis: {age_kurtosis:.4f}")
print(f"Number of Age values within 1 std deviation from the mean: {age_within_1_std}")

print(f"Fare Skewness: {fare_skewness:.4f}")
print(f"Fare Kurtosis: {fare_kurtosis:.4f}")
print(f"Number of Fare values within 1 std deviation from the mean: {fare_within_1_std}")
````

### Summary of Distribution Analysis

#### Age
- **Skewness**: 0.3883
- **Kurtosis**: 0.1686
- **Number of values within 1 standard deviation from the mean**: 516

#### Fare
- **Skewness**: 4.7793
- **Kurtosis**: 33.2043
- **Number of values within 1 standard deviation from the mean**: 818

### Interpretation
- **Age**:
  - The skewness value of 0.3883 indicates a slight right skew, meaning the distribution has a longer tail on the right side.
  - The kurtosis value of 0.1686 suggests the distribution is slightly more peaked than a normal distribution.
  - Approximately 72.27% (516 out of 714) of the 'Age' values fall within one standard deviation from the mean.

- **Fare**:
  - The skewness value of 4.7793 indicates a strong right skew, suggesting a significant number of high fare values.
  - The kurtosis value of 33.2043 indicates a very peaked distribution with heavy tails.
  - Approximately 91.81% (818 out of 891) of the 'Fare' values fall within one standard deviation from the mean.
