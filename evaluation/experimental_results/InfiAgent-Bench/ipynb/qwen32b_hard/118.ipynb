[STEP GOAL]: Perform linear regression to model the relationship between GDP per capita and life expectancy score.


````python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Extract the relevant columns
gdp_per_capita = data['Economy (GDP per Capita)'].values.reshape(-1, 1)
life_expectancy = data['Health (Life Expectancy)'].values.reshape(-1, 1)

# Initialize and fit the linear regression model
model = LinearRegression()
model.fit(gdp_per_capita, life_expectancy)

# Predict the life expectancy scores
predicted_life_expectancy = model.predict(gdp_per_capita)

# Calculate the coefficient of determination (R-squared)
r_squared = r2_score(life_expectancy, predicted_life_expectancy)
r_squared
````

[STEP GOAL]: Evaluate the goodness of fit of the linear regression model based on the R-squared value.


````python
# Define the threshold for a good fit
threshold = 0.7

# Determine if the model is a good fit
if r_squared >= threshold:
    model_fit = "good fit"
else:
    model_fit = "poor fit"

model_fit, r_squared
````

The coefficient of determination (R-squared) for the linear relationship between GDP per capita and life expectancy score is approximately 0.667. Since this value is less than 0.7, we consider the model to be a **poor fit**.
