### Debugging Summary
During the debugging process, we identified that the training data contained NaN values, which caused the `LinearRegression` model to fail. We dropped the rows with NaN values and retrained the model successfully.


````python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Load the dataset
file_path = './input/unemployement_industry.csv'
df = pd.read_csv(file_path)

# Select the relevant columns
X = df[['Mar.2019']]
y = df['Mar.2020']

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Drop rows with NaN values
X_train_clean = X_train.dropna()
y_train_clean = y_train[X_train_clean.index]

# Update the training sets
X_train, y_train = X_train_clean, y_train_clean

# Initialize and train the model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model's performance
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
````

[STEP GOAL]: Evaluate the model's performance on the testing set using Mean Squared Error.


````python
# Calculate and print the Mean Squared Error
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")
````
