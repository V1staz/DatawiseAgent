```text
Question: Explore the distribution of the abs_diffsel column and determine if it adheres to a normal distribution by calculating skewness and kurtosis. The skewness and kurtosis values should be calculated using Fisher’s method. If the skewness value is between -0.5 and 0.5, the data is fairly symmetrical. If the kurtosis value is around 0, then a normal distribution is often assumed.
Perform the calculations using non-parametric methods, specifically the skew and kurtosis functions provided in the scipy.stats module of Python. All numeric values should be rounded to two decimal places.
```

## Expected Format

```text
@skewness_value[float between -0.5 and 0.5, rounded to two decimal places] @kurtosis_value[float around 0, answer should be within range of -0.5 to 0.5, rounded to two decimal places]. Once both calculations are complete, construct a string message that explicitly indicates whether the data is normally distributed.
```


[STEP GOAL]: Load the dataset and check the first few rows to understand its structure. Also, verify the presence of the `abs_diffsel` column.


````python
import pandas as pd

# Load the dataset
file_path = './input/ferret-Pitt-2-preinf-lib2-100_sitediffsel.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
data.head()
````

````python
# Verify the presence of the 'abs_diffsel' column
'abs_diffsel' in data.columns
````

[STEP GOAL]: Calculate the skewness and kurtosis of the `abs_diffsel` column using the `scipy.stats` module and interpret the results.


````python
from scipy.stats import skew, kurtosis

# Calculate skewness and kurtosis
skewness_value = skew(data['abs_diffsel'])
kurtosis_value = kurtosis(data['abs_diffsel'], fisher=True)

# Round the values to two decimal places
skewness_value_rounded = round(skewness_value, 2)
kurtosis_value_rounded = round(kurtosis_value, 2)

skewness_value_rounded, kurtosis_value_rounded
````

[STEP GOAL]: Interpret the skewness and kurtosis values to determine if the `abs_diffsel` column adheres to a normal distribution.


````python
# Interpret the skewness and kurtosis values
skewness_interpretation = "fairly symmetrical" if -0.5 <= skewness_value_rounded <= 0.5 else "not symmetrical"
kurtosis_interpretation = "normal distribution is often assumed" if -0.5 <= kurtosis_value_rounded <= 0.5 else "not normally distributed"

skewness_interpretation, kurtosis_interpretation
````

### Results and Interpretation

The skewness and kurtosis values for the `abs_diffsel` column are as follows:

- **Skewness**: 0.14
- **Kurtosis**: -0.30

#### Interpretation:
- **Skewness**: The skewness value of 0.14 falls within the range of -0.5 to 0.5, indicating that the data is **fairly symmetrical**.
- **Kurtosis**: The kurtosis value of -0.30 is close to 0, suggesting that a **normal distribution is often assumed**.

Based on these values, the `abs_diffsel` column adheres reasonably well to a normal distribution.
