[STEP GOAL]: Remove missing values from the dataset and select the relevant columns for further preprocessing.


````python
# Remove rows with missing values
data_cleaned = data.dropna()

# Select the relevant columns
columns_of_interest = ['Date', 'MSFT.O', 'SPY', '.VIX']
data_cleaned = data_cleaned[columns_of_interest]

# Display the first few rows of the cleaned dataset
data_cleaned.head()
````

[STEP GOAL]: Normalize the selected stock price columns using feature scaling and calculate the correlation matrix using Pearson's correlation.


````python
from sklearn.preprocessing import MinMaxScaler

# Extract the relevant columns for normalization
stock_prices = data_cleaned[['MSFT.O', 'SPY', '.VIX']]

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Fit and transform the data
normalized_stock_prices = scaler.fit_transform(stock_prices)

# Convert the normalized data back to a DataFrame
normalized_stock_prices_df = pd.DataFrame(normalized_stock_prices, columns=['MSFT.O', 'SPY', '.VIX'], index=data_cleaned.index)

# Display the first few rows of the normalized data
normalized_stock_prices_df.head()
````

````python
# Calculate the correlation matrix using Pearson's correlation
correlation_matrix = normalized_stock_prices_df.corr(method='pearson')

# Display the correlation matrix
correlation_matrix
````
