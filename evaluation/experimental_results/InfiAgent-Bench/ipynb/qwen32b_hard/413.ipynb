[STEP GOAL]: Load the Titanic dataset and filter the records for passengers who embarked from Cherbourg (Embarked = 'C').


````python
# Load the Titanic dataset
titanic_df = pd.read_csv('input/titanic_train.csv')

# Filter the dataset for passengers who embarked from Cherbourg (Embarked = 'C')
cherbourg_df = titanic_df[titanic_df['Embarked'] == 'C']

# Display the first few rows of the filtered dataset
cherbourg_df.head()
````

[STEP GOAL]: Calculate the Pearson correlation coefficient (r) and the p-value for the linear relationship between Pclass and Fare for passengers who embarked from Cherbourg.


````python
from scipy.stats import pearsonr

# Calculate the Pearson correlation coefficient and p-value
pclass_fare_corr, p_value = pearsonr(cherbourg_df['Pclass'], cherbourg_df['Fare'])

# Display the results
pclass_fare_corr, p_value
````

### Analysis of Correlation Between Ticket Class (Pclass) and Fare for Passengers Embarking from Cherbourg (Embarked = 'C')

#### Pearson Correlation Coefficient (r):
- **Value**: -0.5307
- **Interpretation**: This indicates a moderate negative linear relationship between the ticket class (Pclass) and the fare paid by the passengers. As the ticket class increases (i.e., lower class), the fare tends to decrease.

#### P-value:
- **Value**: 1.361e-13
- **Significance Level (α)**: 0.01
- **Interpretation**: The p-value is much smaller than the significance level (0.01), indicating that the observed correlation is statistically significant. Therefore, we can reject the null hypothesis that there is no correlation between Pclass and Fare.

#### Conclusion:
There is a significant moderate negative correlation between the ticket class (Pclass) and the fare paid by the passengers who embarked from Cherbourg (Embarked = 'C'). This means that as the ticket class increases, the fare tends to decrease, and this relationship is statistically significant.
