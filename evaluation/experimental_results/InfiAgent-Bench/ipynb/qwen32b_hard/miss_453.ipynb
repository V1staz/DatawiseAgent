````python
# Load the dataset
file_path = './input/baro_2015.csv'
data = pd.read_csv(file_path)

# Ensure there are no leading/trailing spaces in column names
data.columns = data.columns.str.strip()

# Calculate the mean and standard deviation of the wind speed column before preprocessing
mean_windspeed_before = data['WINDSPEED'].mean()
std_windspeed_before = data['WINDSPEED'].std()

# Calculate the Z-scores for the wind speed column
z_scores = (data['WINDSPEED'] - mean_windspeed_before) / std_windspeed_before

# Identify outliers (Z-score > 3 or < -3)
outliers = (z_scores > 3) | (z_scores < -3)

# Replace outliers with the mean wind speed
data.loc[outliers, 'WINDSPEED'] = mean_windspeed_before

# Calculate the mean atmospheric temperature (AT)
mean_at = data['AT'].mean()

# Fill missing values in the atmospheric temperature (AT) column with the mean
data['AT'] = data['AT'].fillna(mean_at)

# Calculate the mean wind speed and average atmospheric temperature after preprocessing
mean_windspeed_after = data['WINDSPEED'].mean()
mean_at_after = data['AT'].mean()

mean_windspeed_before, std_windspeed_before, mean_windspeed_after, mean_at_after
````

[STEP GOAL]: Perform outlier removal on the wind speed (WINDSPEED) column using the Z-score method and handle missing values in the atmospheric temperature (AT) column by replacing them with the mean temperature.


````python
# Calculate the mean and standard deviation of the wind speed column before preprocessing
mean_windspeed_before = data['WINDSPEED'].mean()
std_windspeed_before = data['WINDSPEED'].std()

# Calculate the Z-scores for the wind speed column
z_scores = (data['WINDSPEED'] - mean_windspeed_before) / std_windspeed_before

# Identify outliers (Z-score > 3 or < -3)
outliers = (z_scores > 3) | (z_scores < -3)

# Replace outliers with the mean wind speed
data.loc[outliers, 'WINDSPEED'] = mean_windspeed_before

# Calculate the mean atmospheric temperature (AT)
mean_at = data['AT'].mean()

# Fill missing values in the atmospheric temperature (AT) column with the mean
data['AT'] = data['AT'].fillna(mean_at)

# Calculate the mean wind speed and average atmospheric temperature after preprocessing
mean_windspeed_after = data['WINDSPEED'].mean()
mean_at_after = data['AT'].mean()

mean_windspeed_before, std_windspeed_before, mean_windspeed_after, mean_at_after
````
