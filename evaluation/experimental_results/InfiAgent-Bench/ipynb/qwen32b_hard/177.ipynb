[STEP GOAL]: Filter the dataset to include only passengers from 1st and 3rd classes, and drop rows with missing age values.


````python
# Filter the dataset to include only passengers from 1st and 3rd classes
filtered_df = titanic_df[titanic_df['Pclass'].isin([1, 3])]

# Drop rows with missing age values
filtered_df = filtered_df.dropna(subset=['Age'])

# Display the first few rows of the filtered dataset
filtered_df.head()
````

````python
# Check the number of rows and columns in the filtered dataset
filtered_df.shape
````

The dataset has been successfully filtered to include only passengers from 1st and 3rd classes, and rows with missing age values have been dropped. The filtered dataset now contains 541 rows.


[STEP GOAL]: Extract the age distributions for 1st and 3rd class passengers.


````python
# Extract age distributions for 1st and 3rd class passengers
ages_1st_class = filtered_df[filtered_df['Pclass'] == 1]['Age']
ages_3rd_class = filtered_df[filtered_df['Pclass'] == 3]['Age']

# Display the first few values of each age distribution
ages_1st_class.head(), ages_3rd_class.head()
````

[STEP GOAL]: Perform the Mann-Whitney U test to determine if there is a significant difference in the age distributions between 1st and 3rd class passengers.


````python
from scipy.stats import mannwhitneyu

# Perform the Mann-Whitney U test
stat, p_value = mannwhitneyu(ages_1st_class, ages_3rd_class, alternative='two-sided')

# Display the test statistic and p-value
stat, p_value
````

````python
# Determine if the difference in age distributions is significant
alpha = 0.05
if p_value < alpha:
    result = "There is a significant difference in the age distributions between 1st and 3rd class passengers."
else:
    result = "There is no significant difference in the age distributions between 1st and 3rd class passengers."

result
````

The Mann-Whitney U test has been performed, and the results show a significant difference in the age distributions between 1st and 3rd class passengers at the 0.05 significance level. The p-value is extremely low (8.797424981276622e-22), indicating strong evidence against the null hypothesis.
