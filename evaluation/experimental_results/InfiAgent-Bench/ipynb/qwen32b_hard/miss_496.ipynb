The dataset has been successfully loaded and inspected. It contains 42 entries with 18 columns, representing various fields of study and the percentage of bachelor's degrees awarded to women in the USA over the years. The next step is to create the "STEM" feature.


[STEP GOAL]: Create the "STEM" feature by summing the percentages of graduates in the fields of Computer Science, Engineering, Math and Statistics, and Physical Sciences. Then, calculate the mean and range (maximum - minimum) of the "STEM" feature for the years after 2000, rounding the results to two decimal places.


````python
# Create the "STEM" feature
df['STEM'] = df['Computer Science'] + df['Engineering'] + df['Math and Statistics'] + df['Physical Sciences']

# Filter the dataset for years after 2000
df_post_2000 = df[df['Year'] > 2000]

# Calculate the mean and range of the "STEM" feature for the years after 2000
stem_mean = df_post_2000['STEM'].mean().round(2)
stem_range = (df_post_2000['STEM'].max() - df_post_2000['STEM'].min()).round(2)

stem_mean, stem_range
````

The "STEM" feature has been successfully created by summing the percentages of graduates in the fields of Computer Science, Engineering, Math and Statistics, and Physical Sciences. For the years after 2000, the mean and range (maximum - minimum) of the "STEM" feature are as follows:

- Mean: 124.18%
- Range: 17.70%

These calculations have been rounded to two decimal places.
