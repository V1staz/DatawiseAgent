[STEP GOAL]: Load the dataset and perform initial data exploration to understand its structure and contents.


````python
import pandas as pd

# Load the dataset
file_path = './input/weather_train.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
data.head()
````

````python
# Check the shape of the dataset
data.shape
````

````python
# Check for missing values in the dataset
data.isnull().sum()
````

The dataset has 16,683 rows and 9 columns. It contains the following features:

- `time`: Timestamp of the observation.
- `sun`: Amount of sunlight.
- `temperature`: Temperature in degrees Celsius.
- `chill`: Wind chill.
- `humidity`: Humidity percentage.
- `wind direction`: Direction of the wind.
- `wind speed`: Speed of the wind in meters per second.
- `visibility`: Visibility in meters.
- `air pressure`: Air pressure in hectopascals.

There are some missing values in the dataset:
- `sun`: 52 missing values
- `temperature`: 41 missing values
- `chill`: 8,220 missing values
- `humidity`: 41 missing values
- `wind direction`: 38 missing values
- `wind speed`: 41 missing values
- `visibility`: 11 missing values
- `air pressure`: 39 missing values

We will handle these missing values before training the model.


<end_step>


[STEP GOAL]: Handle missing values in the dataset and prepare the data for model training.


````python
# Drop rows with missing values in the columns we are interested in
data_cleaned = data.dropna(subset=['sun', 'temperature', 'humidity', 'wind speed'])

# Verify the number of rows after dropping missing values
data_cleaned.shape
````

````python
# Display the first few rows of the cleaned dataset
data_cleaned.head()
````

The dataset has been cleaned by removing rows with missing values in the columns of interest (`sun`, `temperature`, `humidity`, `wind speed`). The cleaned dataset now has 16,602 rows.

Next, we will split the dataset into training and testing sets and train a simple linear regression model to predict the amount of sunlight (`sun`) based on the temperature, humidity, and wind speed.


[STEP GOAL]: Split the dataset into training and testing sets, and train a simple linear regression model.


````python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Define the features and target variable
X = data_cleaned[['temperature', 'humidity', 'wind speed']]
y = data_cleaned['sun']

# Split the data into training and testing sets (70% training, 30% testing)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize the linear regression model
model = LinearRegression()

# Train the model
model.fit(X_train, y_train)

# Make predictions on the testing set
y_pred = model.predict(X_test)

# Evaluate the model using mean squared error
mse = mean_squared_error(y_test, y_pred)
mse
````

[STEP GOAL]: Summarize the results and provide the mean squared error of the trained linear regression model.


The simple linear regression model has been trained to predict the amount of sunlight (`sun` column) based on the `temperature`, `humidity`, and `wind speed` columns. The dataset was split into a 70-30 training-testing split. The model's performance was evaluated using the mean squared error (MSE).

The mean squared error of the model on the test set is approximately \(1.179\). This value indicates the average squared difference between the predicted and actual values of sunlight. Since the MSE is positive, it confirms that the model is performing as expected.


````python
# Print the mean squared error
print(f"Mean Squared Error: {mse:.4f}")
````
