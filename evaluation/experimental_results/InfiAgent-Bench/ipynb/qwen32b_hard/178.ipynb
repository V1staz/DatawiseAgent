```text
Question: Perform comprehensive data preprocessing on the dataset. Handle missing values in the "Embarked" column by imputing them with the mode value. Normalize the "Fare" column using Min-Max scaling. Encode the categorical variable "Sex" using Label Encoding, where "male" is coded as 1 and "female" as 0. Calculate the number of each label after processing "Sex" and the minimum, maximum and mean of "Fare" after scaling.
{
"Embarked" missing values should be filled with the mode value.
"Fare" should be normalized using Min-Max scaling, where Min is the minimum value of "Fare" before scaling and Max is the maximum.
"Sex" should be encoded using Label Encoding, where "male" is 1 and "female" is 0.
Caculate the count of each label of "Sex" after encoding and the min, max and mean of "Fare" values after scaling.
}
```

## Expected Format

```text
{
@sex_encoded_count[label_0_count, label_1_count]
@fare_after_scaling[min_fare, max_fare, mean_fare]
where "label_0_count" and "label_1_count" are numbers indicating the count of 0 and 1 in the encoded "Sex" column, respectively.
where "min_fare", "max_fare", and "mean_fare" are the minimum, maximum, and mean of the "Fare" column after Min-Max scaling, respectively. "min_fare" and "max_fare" are rounded to two decimal places, while "mean_fare" is rounded to four decimal places.
```


[STEP GOAL]: Load the dataset and inspect its initial state, including the presence of missing values.


````python
# Load the dataset
import pandas as pd

# Read the CSV file
file_path = './input/titanic.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the dataset
df.head()
````

````python
# Check for missing values in the dataset
df.isnull().sum()
````

[STEP GOAL]: Handle missing values in the "Embarked" column by imputing them with the mode value.


````python
# Impute missing values in the "Embarked" column with the mode value
mode_embarked = df['Embarked'].mode()[0]
df['Embarked'].fillna(mode_embarked, inplace=True)

# Verify that there are no more missing values in the "Embarked" column
df['Embarked'].isnull().sum()
````

[STEP GOAL]: Normalize the "Fare" column using Min-Max scaling.


````python
# Normalize the "Fare" column using Min-Max scaling
from sklearn.preprocessing import MinMaxScaler

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Fit and transform the "Fare" column
df['Fare'] = scaler.fit_transform(df[['Fare']])

# Verify the scaled "Fare" column
df['Fare'].describe()
````

[STEP GOAL]: Encode the categorical variable "Sex" using Label Encoding, where "male" is coded as 1 and "female" as 0.


````python
# Encode the "Sex" column using Label Encoding
df['Sex'] = df['Sex'].map({'male': 1, 'female': 0})

# Verify the encoded "Sex" column
df['Sex'].value_counts()
````

[STEP GOAL]: Calculate the number of each label after processing "Sex" and the minimum, maximum, and mean of "Fare" after scaling.


````python
# Calculate the number of each label after processing "Sex"
sex_counts = df['Sex'].value_counts()

# Calculate the minimum, maximum, and mean of "Fare" after scaling
fare_stats = df['Fare'].agg(['min', 'max', 'mean'])

# Display the results
sex_counts, fare_stats
````

   - The "Fare" column was normalized using Min-Max scaling.

   - The "Sex" column was encoded using Label Encoding, where "male" is coded as 1 and "female" is coded as 0.
