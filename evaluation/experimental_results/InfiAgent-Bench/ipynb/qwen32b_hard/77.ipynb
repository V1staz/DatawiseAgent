## Prompt

```text
Question: Perform comprehensive data preprocessing on the "Date" column to extract the month and year information. Calculate the average closing price for each month and year combination. Return the month and year combination which has the highest average closing price.
Set the start of each month to be the first day of the month and the end of the month to be the last day of the month.
The calculation of the average closing price should be done using the arithmetic mean. 
For ties, return the most recent month and year combination.
```

## Expected Format

```text
@Highest_Monthly_Average_Close_Price[month, year, average_close_price]
where "month" is an integer between 1 and 12, "year" is an integer, and "average_close_price" is a float rounded to two decimal places.
```


[STEP GOAL]: Load the dataset and inspect the initial structure to understand the available columns and data types.


````python
# Load the dataset
df = pd.read_csv('./input/microsoft.csv')

# Display the first few rows to inspect the data
df.head()
````

````python
# Check the data types of the columns
df.dtypes
````

[STEP GOAL]: Convert the 'Date' column to a datetime format and extract the month and year information.


````python
# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%d-%b-%y')

# Extract the month and year from the 'Date' column
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year

# Display the updated dataframe
df.head()
````

[STEP GOAL]: Group the data by the 'Month' and 'Year' columns and calculate the average closing price for each group. Identify the month and year combination with the highest average closing price.


````python
# Group by 'Month' and 'Year' and calculate the average closing price
monthly_avg_close = df.groupby(['Year', 'Month'])['Close'].mean().reset_index()

# Rename the column for clarity
monthly_avg_close.rename(columns={'Close': 'Average_Close'}, inplace=True)

# Display the result
monthly_avg_close.head()
````

````python
# Find the month and year combination with the highest average closing price
max_avg_close = monthly_avg_close.loc[monthly_avg_close['Average_Close'].idxmax()]

# Display the result
max_avg_close
````

The month and year combination with the highest average closing price is January 2018, with an average closing price of $88.32.


[STEP GOAL]: Set the start of each month to be the first day of the month and the end of the month to be the last day of the month.


````python
# Import necessary functions
from pandas.tseries.offsets import MonthBegin, MonthEnd

# Set the start of each month to the first day and the end of each month to the last day
df['Month_Start'] = df['Date'].dt.to_period('M').dt.start_time
df['Month_End'] = df['Date'].dt.to_period('M').dt.end_time

# Display the updated dataframe
df.head()
````

The month and year combination with the highest average closing price is January 2018, with an average closing price of $88.32. The start of this month is set to January 1, 2018, and the end of the month is set to January 31, 2018.


[STEP GOAL]: Identify the month and year combination with the highest average closing price, and handle ties by returning the most recent month and year combination.


````python
# Group by 'Month' and 'Year' and calculate the average closing price
monthly_avg_close = df.groupby(['Year', 'Month'])['Close'].mean().reset_index()

# Rename the column for clarity
monthly_avg_close.rename(columns={'Close': 'Average_Close'}, inplace=True)

# Sort the dataframe by 'Year' and 'Month' in descending order to handle ties by returning the most recent month and year combination
monthly_avg_close.sort_values(by=['Year', 'Month'], ascending=False, inplace=True)

# Find the month and year combination with the highest average closing price
max_avg_close = monthly_avg_close.loc[monthly_avg_close['Average_Close'].idxmax()]

# Display the result
max_avg_close
````

The month and year combination with the highest average closing price is January 2018, with an average closing price of $88.32. In case of ties, the most recent month and year combination is returned, which is also January 2018 in this scenario.
