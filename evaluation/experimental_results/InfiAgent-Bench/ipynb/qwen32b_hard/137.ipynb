[STEP GOAL]: Create the new binary feature "IsAlone" which indicates whether a passenger is traveling alone or with family. This will be determined by checking if both "SibSp" and "Parch" are zero.


````python
# Create the 'IsAlone' feature
titanic_df['IsAlone'] = (titanic_df['SibSp'] + titanic_df['Parch'] == 0).astype(int)

# Display the first few rows of the updated dataset
titanic_df[['PassengerId', 'Survived', 'SibSp', 'Parch', 'IsAlone']].head()
````

[STEP GOAL]: Split the dataset into training and testing sets using a 70:30 ratio. Ensure the class weights are balanced and set the random seed to 42 for reproducibility.


````python
from sklearn.model_selection import train_test_split

# Select the features and target variable
X = titanic_df[['IsAlone']]
y = titanic_df['Survived']

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, stratify=y, random_state=42)

# Display the shapes of the resulting datasets
X_train.shape, X_test.shape, y_train.shape, y_test.shape
````

The dataset has been successfully split into training and testing sets with a 70:30 ratio, ensuring that the class weights are balanced. The random seed has been set to 42 for reproducibility.


[STEP GOAL]: Train a logistic regression model using the 'IsAlone' feature and the 'Survived' column as the target variable. Use scikit-learn's LogisticRegression with default parameters and balance the class weights.


````python
from sklearn.linear_model import LogisticRegression

# Initialize the logistic regression model with balanced class weights
log_reg = LogisticRegression(class_weight='balanced', random_state=42)

# Train the model
log_reg.fit(X_train, y_train)
````

[STEP GOAL]: Evaluate the performance of the trained logistic regression model on the test set using accuracy and confusion matrix.


````python
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Make predictions on the test set
y_pred = log_reg.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')

# Generate confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Display confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=log_reg.classes_)
disp.plot()
plt.show()
````
