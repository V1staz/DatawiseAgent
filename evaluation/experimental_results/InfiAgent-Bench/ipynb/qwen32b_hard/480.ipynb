[STEP GOAL]: Create a new feature by subtracting the mean value of the "Value" column from each value in that column. Then, calculate the standard deviation of this new feature.


````python
# Calculate the mean of the "Value" column
mean_value = df['Value'].mean()

# Create a new feature by subtracting the mean value from each value in the "Value" column
df['Value_centered'] = df['Value'] - mean_value

# Calculate the standard deviation of the new feature
std_dev_value_centered = df['Value_centered'].std()

# Display the result
mean_value, std_dev_value_centered
````
