```text
Question: Identify the date with the highest closing value of the S&P 500 Index (.SPX). Calculate the percentage change in the stock price of Apple Inc. (AAPL) from its closing price on the previous day to its closing price on the identified date.
1. The date where the S&P 500 Index (.SPX) reached its maximum value should be identified.
2. The percentage change is calculated as: ((price on identified date / price on previous day) - 1) * 100.
3. Percentage change should be calculated only if the previous day data exists. If the identified date is the first date in the dataset, state that the previous day data doesn't exist.
4. The data for the previous day is defined as the data on the date immediately preceding the identified date when sorting the dates in ascending order. Hunting for the "previous" trading day is not required.
```

## Expected Format

```text
@max_SPX_date[date]
@AAPL_price_percentage_change[percentage_change]

```


[STEP GOAL]: Load the provided CSV file into a DataFrame and inspect the first few rows to understand the structure of the dataset.


````python
# Load the CSV file into a DataFrame
file_path = './input/tr_eikon_eod_data.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame
df.head()
````

[STEP GOAL]: Identify the date with the highest closing value of the S&P 500 Index (.SPX).

We will find the maximum value of the `.SPX` column and the corresponding date.


````python
# Find the date with the highest closing value of the S&P 500 Index (.SPX)
max_spx_date = df.loc[df['.SPX'].idxmax(), 'Date']
max_spx_value = df['.SPX'].max()

max_spx_date, max_spx_value
````

[STEP GOAL]: Calculate the percentage change in the stock price of Apple Inc. (AAPL) from its closing price on the previous day to its closing price on the identified date.


````python
# Extract the closing price of Apple Inc. (AAPL) on the identified date
identified_date_aapl_price = df.loc[df['Date'] == max_spx_date, 'AAPL.O'].values[0]

# Find the previous day's data
previous_day_df = df[df['Date'] < max_spx_date].sort_values(by='Date', ascending=False).head(1)

# Check if the previous day's data exists
if not previous_day_df.empty:
    previous_day_aapl_price = previous_day_df['AAPL.O'].values[0]
    
    # Calculate the percentage change
    percentage_change = ((identified_date_aapl_price / previous_day_aapl_price) - 1) * 100
else:
    percentage_change = None

identified_date_aapl_price, previous_day_aapl_price, percentage_change
````

The date with the highest closing value of the S&P 500 Index (.SPX) is **2018-01-26** with a closing value of **2872.87**.

The percentage change in the stock price of Apple Inc. (AAPL) from its closing price on the previous day (2018-01-25) to its closing price on the identified date (2018-01-26) is **0.23%**.
