### Summary of Preprocessing Steps

1. **Loaded the Dataset**: Read the CSV file and inspected the initial structure and summary statistics.
2. **Transformed Categorical Columns**: Mapped the 'sex' and 'smoker' columns to binary format (0 and 1).
3. **Normalized Numeric Columns**: Applied MinMaxScaler to the 'age', 'bmi', 'children', and 'charges' columns.

````python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Load the dataset
file_path = './input/insurance.csv'
df = pd.read_csv(file_path)

# Map 'sex' and 'smoker' columns to binary format
df['sex'] = df['sex'].map({'female': 0, 'male': 1})
df['smoker'] = df['smoker'].map({'no': 0, 'yes': 1})

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Columns to normalize
columns_to_normalize = ['age', 'bmi', 'children', 'charges']

# Apply normalization
df[columns_to_normalize] = scaler.fit_transform(df[columns_to_normalize])

# Convert the 'region' column to a category type
df['region'] = df['region'].astype('category')

# Calculate and report the mean of each column after preprocessing
mean_values = df.mean(numeric_only=True)
print(mean_values)

# Save the preprocessed dataset for future use
preprocessed_file_path = './working/preprocessed_insurance.csv'
df.to_csv(preprocessed_file_path, index=False)
````

1. **Handling Missing Values**: No missing values were found in the dataset.
2. **Binary Transformation**:
   - 'sex': Female -> 0, Male -> 1
   - 'smoker': No -> 0, Yes -> 1
3. **Normalization**:
   - 'age', 'bmi', 'children', and 'charges' were normalized using MinMaxScaler.
