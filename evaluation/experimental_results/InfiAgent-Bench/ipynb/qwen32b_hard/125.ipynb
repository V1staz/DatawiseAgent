The dataset contains the required columns: 'total_vaccinations', 'people_vaccinated_per_hundred', and 'people_fully_vaccinated_per_hundred'. However, there are missing values in these columns. We will filter out the rows with missing values before proceeding with the analysis.


````python
# Filter out rows with missing values in the required columns
df_clean = df.dropna(subset=['total_vaccinations', 'people_vaccinated_per_hundred', 'people_fully_vaccinated_per_hundred'])

# Display the first few rows of the cleaned dataset
df_clean.head()
````

[STEP GOAL]: Perform multiple linear regression using the cleaned dataset. The independent variables are 'total_vaccinations' and 'people_vaccinated_per_hundred', and the dependent variable is 'people_fully_vaccinated_per_hundred'.


````python
# Import necessary libraries
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm

# Define the independent and dependent variables
X = df_clean[['total_vaccinations', 'people_vaccinated_per_hundred']]
y = df_clean['people_fully_vaccinated_per_hundred']

# Fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# Output the coefficients and intercept
print(f'Intercept: {model.intercept_}')
print(f'Coefficients: {model.coef_}')
````

````python
# Add a constant term to the independent variables for the statsmodels model
X_sm = sm.add_constant(X)

# Fit the OLS model
model_sm = sm.OLS(y, X_sm).fit()

# Print the summary of the model
print(model_sm.summary())
````

[STEP GOAL]: Interpret the results of the multiple linear regression analysis, including the coefficients, p-values, and R-squared value.
