[STEP GOAL]: Preprocess the data by encoding the gender column and handling missing values in the age column.


````python
# Check for missing values
titanic_df.isnull().sum()
````

````python
# Encode the 'Sex' column
titanic_df['Sex'] = titanic_df['Sex'].map({'male': 0, 'female': 1})

# Handle missing values in the 'Age' column by imputing with the median
titanic_df['Age'].fillna(titanic_df['Age'].median(), inplace=True)

# Display the first few rows of the preprocessed dataset
titanic_df.head()
````

The data has been preprocessed:
- The `Sex` column has been encoded as `0` for male and `1` for female.
- Missing values in the `Age` column have been imputed with the median age.

Next, we will split the dataset into training and test sets and prepare the features and target variable.


[STEP GOAL]: Split the dataset into training and test sets with a 70:30 ratio and prepare the features and target variable.


````python
from sklearn.model_selection import train_test_split

# Define features and target variable
features = ['Age', 'Sex', 'Fare']
X = titanic_df[features]
y = titanic_df['Survived']

# Split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Display the shapes of the training and test sets
X_train.shape, X_test.shape, y_train.shape, y_test.shape
````

[STEP GOAL]: Train a logistic regression model with default parameters and evaluate its performance using accuracy as the evaluation metric.


````python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Initialize the logistic regression model with default parameters
model = LogisticRegression()

# Train the model
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Calculate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)

# Display the accuracy
accuracy
````

The logistic regression model has been trained and evaluated. The accuracy of the model on the test set is approximately 78.36%.

This concludes the task of building and evaluating a logistic regression model to predict survival on the Titanic dataset based on age, gender, and fare.
