```text
Question: Create a new feature called "Price Range" which represents the difference between the "High Price" and "Low Price" for each row. Calculate the mean, median, and standard deviation of this new feature.
Make sure to use the correct columns for calculating the "Price Range". All calculations should be performed up to two decimal places.
```

## Expected Format

```text
@price_range_mean[mean]: The mean should be a single real number rounded to two decimal places. @price_range_median[median]: The median should be a single real number rounded to two decimal places. @price_range_std_dev[std_dev]: The standard deviation should be a single real number rounded to two decimal places.
```


[STEP GOAL]: Load the dataset from the CSV file and create a new feature "Price Range" based on the difference between the "High Price" and "Low Price".


````python
# Load the dataset
df = pd.read_csv('./input/GODREJIND.csv')

# Display the first few rows to understand the structure
df.head()
````

[STEP GOAL]: Calculate the mean, median, and standard deviation of the "Price Range" and format the results to two decimal places.


````python
# Calculate the price range
df['Price Range'] = df['High Price'] - df['Low Price']

# Calculate mean, median, and standard deviation of the "Price Range"
mean_price_range = round(df['Price Range'].mean(), 2)
median_price_range = round(df['Price Range'].median(), 2)
std_price_range = round(df['Price Range'].std(), 2)

(mean_price_range, median_price_range, std_price_range)
````

The mean, median, and standard deviation of the "Price Range" have been calculated and are as follows:

- Mean Price Range: 16.65
- Median Price Range: 15.67
- Standard Deviation of Price Range: 6.72

These values are rounded to two decimal places as requested.
