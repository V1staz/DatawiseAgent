````python
from scipy.stats import zscore

# Load the Titanic dataset
titanic_df = pd.read_csv('./input/titanic.csv')

# Group the dataframe by 'Pclass' and calculate the Z-scores for the 'Fare' column for each group
grouped = titanic_df.groupby('Pclass')['Fare'].transform(zscore)

# Add the Z-scores as a new column to the original dataframe
titanic_df['zscore_Fare'] = grouped

# Identify outliers (Z-score > 3 or < -3)
outliers = titanic_df[(titanic_df['zscore_Fare'] > 3) | (titanic_df['zscore_Fare'] < -3)]

# Count the number of outliers per passenger class
outliers_count = outliers['Pclass'].value_counts().sort_index()

# Display the count of outliers per passenger class
outliers_count
````

- First Class (Pclass=1): 3 outliers
- Second Class (Pclass=2): 7 outliers
- Third Class (Pclass=3): 14 outliers

This concludes the identification of outliers for the fare variable in the Titanic dataset, separated by passenger class.
