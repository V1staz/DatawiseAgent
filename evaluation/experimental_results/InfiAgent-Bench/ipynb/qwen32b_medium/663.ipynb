[STEP GOAL]: Create a scatter plot of the 'High' and 'Low' columns to visualize the relationship between the highest and lowest prices for each day.


````python
import matplotlib.pyplot as plt

# Create a scatter plot of the 'High' and 'Low' columns
plt.figure(figsize=(10, 6))
plt.scatter(df['High'], df['Low'])
plt.xlabel('High Price')
plt.ylabel('Low Price')
plt.title('Scatter Plot of High vs Low Prices')
plt.show()
````

[STEP GOAL]: Calculate the Pearson correlation coefficient between the 'High' and 'Low' columns, round it to two decimal places, and print the result.


````python
# Calculate the Pearson correlation coefficient between 'High' and 'Low'
correlation_coefficient = df['High'].corr(df['Low'], method='pearson')

# Round the correlation coefficient to two decimal places
correlation_coefficient_rounded = round(correlation_coefficient, 2)

print(f"Pearson correlation coefficient between 'High' and 'Low': {correlation_coefficient_rounded}")
````

The Pearson correlation coefficient between the 'High' and 'Low' columns is 1.00, indicating a perfect positive correlation between the highest and lowest prices for each day. This suggests that as the high price increases, the low price also tends to increase in a linear fashion. However, given the nature of stock prices, this seems unusual and may warrant further investigation into the data integrity or the presence of outliers.
