```text
Question: 3. Using feature engineering, create a new feature called "time_of_day" based on the "dt" column. The "time_of_day" feature should categorize the timestamp into morning (6:00 to 11:59), afternoon (12:00 to 17:59), evening (18:00 to 23:59), and night (0:00 to 5:59) (included). Provide the count of each category in the "time_of_day" column.
For each time of the day, include the first minute of each category and exclude the first minute of the next category. If there's multiple entry which belongs to the same minute, account them all into the corresponding category.
```

## Expected Format

```text
@morning[integer], @afternoon[integer], @evening[integer], @night[integer]
```


[STEP GOAL]: Load the dataset and inspect the initial structure, especially the 'dt' column which contains the timestamps.


````python
# Load the CSV file into a DataFrame
file_path = './input/ravenna_250715.csv'
df = pd.read_csv(file_path)

# Display the first few rows to inspect the structure
df.head()
````

The dataset has been loaded successfully. The 'dt' column appears to contain Unix timestamp values. We will convert these timestamps to a more readable datetime format and then proceed to create the 'time_of_day' feature.


````python
# Convert the 'dt' column to datetime
df['dt'] = pd.to_datetime(df['dt'], unit='s')

# Display the first few rows to verify the conversion
df.head()
````

[STEP GOAL]: Create the 'time_of_day' feature based on the 'dt' column and categorize the timestamps into morning, afternoon, evening, and night. Then, provide the count of each category.


````python
# Define the function to categorize the time of day
def categorize_time_of_day(timestamp):
    hour = timestamp.hour
    if 6 <= hour < 12:
        return 'morning'
    elif 12 <= hour < 18:
        return 'afternoon'
    elif 18 <= hour < 24:
        return 'evening'
    else:
        return 'night'

# Apply the function to the 'dt' column to create the 'time_of_day' feature
df['time_of_day'] = df['dt'].apply(categorize_time_of_day)

# Display the first few rows to verify the new feature
df.head()
````

````python
# Count the occurrences of each time of day category
time_of_day_counts = df['time_of_day'].value_counts()

# Display the counts
time_of_day_counts
````

The 'time_of_day' feature has been created and categorized based on the 'dt' column. Here are the counts for each category:

- Morning: 6
- Afternoon: 6
- Evening: 6
- Night: 6

This indicates that the dataset is evenly distributed across the four time periods.
