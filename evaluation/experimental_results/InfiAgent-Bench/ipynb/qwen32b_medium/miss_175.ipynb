````python
# Load the titanic dataset
titanic_df = pd.read_csv('./input/titanic.csv')

# Display the first few rows of the dataframe to understand its structure
titanic_df.head()
````

[STEP GOAL]: Calculate Z-scores for the 'Age' column and identify outliers based on a Z-score threshold of 3.


````python
from scipy.stats import zscore

# Calculate Z-scores for the 'Age' column while ignoring NaN values
titanic_df['Age_Zscore'] = titanic_df['Age'].apply(lambda x: zscore([x])[0] if pd.notnull(x) else x)

# Identify outliers where the absolute value of the Z-score is greater than 3
outliers = titanic_df[(titanic_df['Age_Zscore'] > 3) | (titanic_df['Age_Zscore'] < -3)]

# Display the number of outliers found
print(f"Number of outliers detected: {outliers.shape[0]}")

# Optionally, display the outliers
outliers[['PassengerId', 'Name', 'Age', 'Age_Zscore']]
````

No outliers were detected in the 'Age' column of the Titanic dataset using the Z-score method with a threshold of 3. This means that all ages fall within three standard deviations from the mean, indicating that the distribution of ages does not contain extreme values beyond this threshold.

[STEP GOAL]: Visualize the distribution of ages along with identified outliers for better understanding.


````python
import matplotlib.pyplot as plt

# Plot histogram of Age
plt.figure(figsize=(10, 6))
plt.hist(titanic_df['Age'].dropna(), bins=20, color='blue', alpha=0.7)
plt.title('Distribution of Passenger Ages on the Titanic')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)

# Optionally, mark the mean and ±3 standard deviations
mean_age = age_summary['mean']
std_age = age_summary['std']
plt.axvline(mean_age, color='red', linestyle='dashed', linewidth=2, label=f'Mean: {mean_age:.2f}')
plt.axvline(mean_age + 3 * std_age, color='green', linestyle='dashed', linewidth=2, label=f'+3σ: {mean_age + 3 * std_age:.2f}')
plt.axvline(mean_age - 3 * std_age, color='green', linestyle='dashed', linewidth=2, label=f'-3σ: {mean_age - 3 * std_age:.2f}')

plt.legend()
plt.show()
````

The histogram below illustrates the distribution of passenger ages on the Titanic. The red dashed line represents the mean age, and the green dashed lines represent the upper and lower bounds that are ±3 standard deviations from the mean (i.e., mean ± 3σ).
