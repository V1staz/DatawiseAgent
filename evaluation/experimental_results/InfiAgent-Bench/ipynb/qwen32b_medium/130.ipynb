[STEP GOAL]: Conduct the Shapiro-Wilk test on the 'Age' column of the titanic dataset, ignoring null values, to check if the data is normally distributed.


````python
from scipy.stats import shapiro

# Select the 'Age' column and remove null values
age_data = df_titanic['Age'].dropna()

# Perform Shapiro-Wilk test to check for normal distribution
stat, p_value = shapiro(age_data)

print(f"Shapiro-Wilk Statistic: {stat}")
print(f"P Value: {p_value}")

# Interpretation based on p-value
alpha = 0.05
if p_value > alpha:
    print("The null hypothesis cannot be rejected. There is evidence that the data is normally distributed.")
else:
    print("The null hypothesis is rejected. There is significant evidence that the data is not normally distributed.")
````

The Shapiro-Wilk test has been conducted on the 'Age' column of the Titanic dataset, excluding any null values. The test resulted in a p-value of approximately 7.34e-08, which is much smaller than the significance level of 0.05. 

Therefore, we can conclude that the null hypothesis, which states the ages are normally distributed, is rejected. There is significant evidence to suggest that the ages of the passengers on the Titanic do not follow a normal distribution.


The Shapiro-Wilk test has been successfully carried out on the 'Age' column from the Titanic dataset, ensuring to exclude any records with missing 'Age' values. The test returned a Shapiro-Wilk statistic of approximately 0.981, and a p-value of approximately 7.34e-08.

Given that the p-value is significantly less than the threshold of 0.05, we reject the null hypothesis. This implies strong evidence against the 'Age' data following a normal distribution. Therefore, the ages of the passengers on the Titanic do not appear to be normally distributed according to statistical testing.
