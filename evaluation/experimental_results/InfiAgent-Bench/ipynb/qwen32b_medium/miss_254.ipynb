[STEP GOAL]: Load the dataset from the input directory and extract the GDP per capita data for the year 1982.


````python
# Loading the dataset
file_path = "./input/gapminder_gdp_asia.csv"
df = pd.read_csv(file_path)

# Displaying the first few rows of the dataframe to understand its structure
df.head()
````

The dataset has been successfully loaded, and we can see that the column `gdpPercap_1982` contains the GDP per capita data for the year 1982.


[STEP GOAL]: Calculate the IQR and identify outliers in the gdpPercap_1982 data.


````python
# Extracting the gdpPercap_1982 column
gdp_1982 = df['gdpPercap_1982']

# Calculating the first quartile (Q1), third quartile (Q3), and interquartile range (IQR)
Q1 = gdp_1982.quantile(0.25)
Q3 = gdp_1982.quantile(0.75)
IQR = Q3 - Q1

# Defining the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identifying outliers
outliers = df[(gdp_1982 < lower_bound) | (gdp_1982 > upper_bound)]

outliers[['country', 'gdpPercap_1982']]
````

The outliers in the GDP per capita data for the year 1982 based on the IQR rule are identified as Kuwait and Saudi Arabia. Their GDP per capita values for 1982 exceed the upper bound established by the IQR rule, indicating they are outliers.
