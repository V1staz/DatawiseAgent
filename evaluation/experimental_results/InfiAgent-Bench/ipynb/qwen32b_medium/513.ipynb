[STEP GOAL]: Load the hotel data from `hotel_data.csv`, and filter out rows where star ratings are null.


````python
# Load the hotel data
df_hotels = pd.read_csv('./input/hotel_data.csv')

# Display the first few rows of the dataframe to understand the structure
df_hotels.head()
````

[STEP GOAL]: Filter the dataframe based on star rating categories: below 3 stars, between 3 and 4 stars, and above 4 stars.


````python
# Filter the dataframe for hotels with non-null star ratings
df_non_null_stars = df_hotels.dropna(subset=['star_rating'])

# Define the groups for star ratings
below_3_stars = df_non_null_stars[df_non_null_stars['star_rating'] < 3]
between_3_and_4_stars = df_non_null_stars[(df_non_null_stars['star_rating'] >= 3) & (df_non_null_stars['star_rating'] <= 4)]
above_4_stars = df_non_null_stars[df_non_null_stars['star_rating'] > 4]

# Display the counts of each category
print(f"Hotels with less than 3 stars: {len(below_3_stars)}")
print(f"Hotels with 3 to 4 stars: {len(between_3_and_4_stars)}")
print(f"Hotels with more than 4 stars: {len(above_4_stars)}")
````

We have successfully filtered the dataframe into three categories based on the star rating of the hotels: below 3 stars, between 3 and 4 stars, and above 4 stars. Now let's calculate the Pearson correlation coefficients for each group between the number of reviews a hotel has received and its bubble score.


[STEP GOAL]: Calculate the Pearson correlation coefficient for each star rating category between the number of reviews and the bubble score.


[STEP GOAL]: Calculate the Pearson correlation coefficient for each star rating category between the number of reviews and the bubble score.


````python
# Calculate the Pearson correlation coefficients
correlation_below_3 = below_3_stars[['review_count', 'bubble_score']].corr(method='pearson').iloc[0, 1]
correlation_between_3_and_4 = between_3_and_4_stars[['review_count', 'bubble_score']].corr(method='pearson').iloc[0, 1]
correlation_above_4 = above_4_stars[['review_count', 'bubble_score']].corr(method='pearson').iloc[0, 1]

# Print the correlations
print(f"Pearson correlation coefficient for hotels with less than 3 stars: {correlation_below_3}")
print(f"Pearson correlation coefficient for hotels with 3 to 4 stars: {correlation_between_3_and_4}")
print(f"Pearson correlation coefficient for hotels with more than 4 stars: {correlation_above_4}")
````

The Pearson correlation coefficients for each star rating category between the number of reviews and the bubble score have been calculated. Here are the results:

- For hotels with less than 3 stars, the correlation coefficient is approximately 0.149, indicating a weak positive relationship between the number of reviews and the bubble score.
- For hotels with 3 to 4 stars, the correlation coefficient is approximately 0.044, indicating a very weak positive relationship.
- For hotels with more than 4 stars, the correlation coefficient is approximately -0.275, indicating a weak negative relationship between the number of reviews and the bubble score.

This suggests that for lower-rated hotels, there may be a slight tendency for those with more reviews to have higher bubble scores, but this trend reverses for the highest-rated hotels.
