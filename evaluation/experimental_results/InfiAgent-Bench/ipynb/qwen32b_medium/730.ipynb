[STEP GOAL]: Load the gapminder_cleaned.csv dataset and calculate the Pearson correlation coefficient between "Pop" and "Gdppercap".


The initial error was due to incorrect column name references. The correct column names are 'pop' and 'gdppercap' instead of 'Pop' and 'Gdppercap'.


````python
import pandas as pd
from scipy.stats import pearsonr

# Load the dataset
data_path = './input/gapminder_cleaned.csv'
gapminder_df = pd.read_csv(data_path)

# Calculate the Pearson correlation coefficient between "pop" and "gdppercap"
correlation, p_value = pearsonr(gapminder_df['pop'], gapminder_df['gdppercap'])

correlation, p_value
````

The Pearson correlation coefficient between "pop" and "gdppercap" is approximately -0.026 with a p-value of 0.291. Since the p-value is greater than 0.05, we do not have sufficient evidence to conclude that there is a statistically significant linear relationship between population and GDP per capita in the dataset.
