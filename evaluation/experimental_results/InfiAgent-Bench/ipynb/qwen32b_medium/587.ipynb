````python
# Load the dataset
df = pd.read_csv('./input/20170413_000000_group_statistics.csv')

# Convert 'avg. wait time ' from 'HH:MM:SS' format to seconds
def convert_to_seconds(time_str):
    h, m, s = map(int, time_str.split(':'))
    return h * 3600 + m * 60 + s

df['avg. wait time '] = df['avg. wait time '].apply(convert_to_seconds)

# Calculate the Pearson's correlation coefficient
correlation_coefficient = df['avg. wait time '].corr(df['avg. num. agents talking '], method='pearson')

# Round the result to three decimal places
correlation_coefficient_rounded = round(correlation_coefficient, 3)

print(f"Pearson's Correlation Coefficient: {correlation_coefficient_rounded}")
````

The average waiting time has been successfully converted to seconds, and the Pearson's correlation coefficient between the average number of agents talking and the transformed average waiting time is 0.639. This indicates a moderate positive correlation between these two variables.
