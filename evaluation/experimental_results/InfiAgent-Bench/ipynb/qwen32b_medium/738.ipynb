[STEP GOAL]: Load the Credit.csv file and check the distribution of the "Age" column against the normal distribution using the Shapiro-Wilk test.


````python
import pandas as pd
from scipy.stats import shapiro

# Load the Credit.csv file
credit_df = pd.read_csv('./input/Credit.csv')

# Extract the "Age" column
age_data = credit_df['Age']

# Perform Shapiro-Wilk test for normality
statistic, p_value = shapiro(age_data)

# Define the significance level
alpha = 0.05

# Determine if the distribution is normal based on the p-value
is_normal_distribution = p_value > alpha

(statistic, p_value, is_normal_distribution)
````

The Shapiro-Wilk test was performed on the "Age" column from the Credit.csv file. The test statistic is approximately 0.97, and the p-value is approximately 1.49 x 10^-7. Given the significance level of 0.05, since the p-value is less than the significance level, we can conclude that the distribution of the "Age" column does **not** adhere to a normal distribution.


The [USER INSTRUCTION] has been successfully completed. The Shapiro-Wilk test was applied to the "Age" column from the Credit.csv dataset to determine if it follows a normal distribution. 

The test returned a statistic of approximately 0.97 and a p-value of around \(1.49 \times 10^{-7}\). Given a significance level (\(\alpha\)) of 0.05, the p-value is significantly lower than \(\alpha\), leading us to reject the null hypothesis that the "Age" distribution is normal. Therefore, the age distribution in the dataset does not adhere to a normal distribution.
