[STEP GOAL]: Load the weather data from `weather_data_1864.csv` and check for missing values in the `obs_value` column for each station.


````python
# Load the weather data from the csv file
weather_df = pd.read_csv('./input/weather_data_1864.csv')

# Display the first few rows of the DataFrame to understand its structure
weather_df.head()
````

[STEP GOAL]: Calculate the number of missing values for each station and identify the station with the most missing values.


````python
# Calculate the number of missing values in the obs_value column for each station
missing_values_by_station = weather_df.groupby('station')['obs_value'].apply(lambda x: x.isna().sum())

# Find the station with the most missing values and its count
station_with_most_missing_values = missing_values_by_station.idxmax()
count_of_missing_values = missing_values_by_station.max()

missing_values_by_station, station_with_most_missing_values, count_of_missing_values
````
