````python
# Function to detect outliers using IQR method for each group
def detect_outliers_iqr(df, column_name):
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column_name] < lower_bound) | (df[column_name] > upper_bound)].index

# Apply the function for each country to find outliers
outliers_indices = []
for country, group in data.groupby('Country'):
    outliers_country = detect_outliers_iqr(group, 'No. of deaths_max')
    outliers_indices.extend(outliers_country)

outliers_indices = list(set(outliers_indices))

# Remove outliers and calculate the mean number of deaths without outliers
data_cleaned = data.drop(index=outliers_indices)
mean_deaths_no_outliers = data_cleaned['No. of deaths_max'].mean()
mean_deaths_no_outliers
````
