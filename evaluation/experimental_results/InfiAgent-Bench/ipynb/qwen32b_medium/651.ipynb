[STEP GOAL]: Calculate the interquartile range (IQR) and identify the upper and lower bounds for outliers based on the Z-coordinate column.


````python
# Calculate the IQR for the Z-coordinate column
Q1 = df['Z'].quantile(0.25)
Q3 = df['Z'].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f"Lower Bound: {lower_bound}, Upper Bound: {upper_bound}")
````

````python
# Identify the outliers in the Z-coordinate column
outliers_below = df[df['Z'] < lower_bound]
outliers_above = df[df['Z'] > upper_bound]

# Count the total number of outliers
num_outliers = len(outliers_below) + len(outliers_above)
print(f"Number of outliers: {num_outliers}")
````
