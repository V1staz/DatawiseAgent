Question: Create a new feature called "FamilySize" by adding the "SibSp" and "Parch" columns together. What is the mean "FamilySize" for passengers who survived versus passengers who did not survive?
Calculate the mean of "FamilySize" separately for the passengers who survived and the passengers who did not survive. "FamilySize" should be an integer value. The mean should be calculated rounding up to two decimal places.
```

```


[STEP GOAL]: Load the titanic_train.csv dataset and create a new feature called "FamilySize" by adding the "SibSp" and "Parch" columns together.


````python
# Load the titanic_train.csv dataset
df_titanic = pd.read_csv('./input/titanic_train.csv')

# Create a new feature called "FamilySize"
df_titanic['FamilySize'] = df_titanic['SibSp'] + df_titanic['Parch']
````

[STEP GOAL]: Calculate the mean "FamilySize" for passengers who survived versus passengers who did not survive, rounding up to two decimal places.


````python
# Calculate the mean "FamilySize" for survivors and non-survivors
mean_family_size_survived = df_titanic[df_titanic['Survived'] == 1]['FamilySize'].mean().round(2)
mean_family_size_not_survived = df_titanic[df_titanic['Survived'] == 0]['FamilySize'].mean().round(2)

mean_family_size_survived, mean_family_size_not_survived
````

The mean "FamilySize" for passengers who survived is 0.94, while for those who did not survive it is 0.88.
