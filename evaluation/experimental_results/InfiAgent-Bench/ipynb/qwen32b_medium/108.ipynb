[STEP GOAL]: Load the dataset and generate the new feature "TotalIncome".


First, we will load the dataset and create a new feature named "TotalIncome" by summing up the values of "ApplicantIncome" and "CoapplicantIncome".


````python
# Load the dataset
data_path = './input/test_Y3wMUE5_7gLdaTN.csv'
df = pd.read_csv(data_path)

# Generate the new feature "TotalIncome"
df['TotalIncome'] = df['ApplicantIncome'] + df['CoapplicantIncome']

# Display the first few rows of the dataframe to verify the new feature
df.head()
````

[STEP GOAL]: Calculate the mean and standard deviation of the "TotalIncome" column and round the results to two decimal places.


Now, we calculate the mean and standard deviation of the "TotalIncome" column and round the results to two decimal places.


````python
mean_total_income = df['TotalIncome'].mean().round(2)
std_total_income = df['TotalIncome'].std().round(2)

mean_total_income, std_total_income
````
