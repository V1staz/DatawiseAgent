[STEP GOAL]: Perform distribution analysis on the "Trips over the past 24-hours (midnight to 11:59pm)" column. This includes checking for normality using the Shapiro-Wilk test, and calculating skewness and kurtosis to determine if the distribution has heavy tails or is skewed.


````python
from scipy.stats import shapiro, skew, kurtosis

# Extract the column of interest
trips_column = df['Trips over the past 24-hours (midnight to 11:59pm)']

# Perform Shapiro-Wilk test for normality
shapiro_test_result = shapiro(trips_column.dropna())

# Calculate skewness and kurtosis
skewness = skew(trips_column.dropna())
kurt = kurtosis(trips_column.dropna())

# Print results
print(f"Shapiro-Wilk Test Statistic: {shapiro_test_result.statistic}, p-value: {shapiro_test_result.pvalue}")
print(f"Skewness: {skewness}")
print(f"Kurtosis: {kurt}")
````

The Shapiro-Wilk test statistic is 0.954, and the p-value is 0.0027. Since the p-value is less than a commonly used significance level of 0.05, we reject the null hypothesis that the data was drawn from a normal distribution. Therefore, the data does not adhere to a normal distribution.

Regarding skewness, the value is approximately 0.152, which is within the range of -0.5 to 0.5, indicating that the distribution is nearly symmetrical.

For kurtosis, the value is around -1.134, which falls outside the range of -0.5 to 0.5. This suggests that the distribution has lighter tails than a normal distribution (platykurtic).


The distribution analysis on the "Trips over the past 24-hours (midnight to 11:59pm)" column has been completed:

- **Normality Check**: According to the Shapiro-Wilk test, the p-value is 0.0027, which is less than 0.05. Therefore, we reject the null hypothesis, suggesting that the data does not follow a normal distribution.
- **Skewness**: The skewness is approximately 0.152, falling within the range of -0.5 to 0.5, indicating that the distribution is nearly symmetrical.
- **Kurtosis**: The kurtosis is around -1.134, which is outside the range of -0.5 to 0.5, indicating that the distribution has lighter tails than a normal distribution (platykurtic).

This concludes the distribution analysis as per the [USER INSTRUCTION].
