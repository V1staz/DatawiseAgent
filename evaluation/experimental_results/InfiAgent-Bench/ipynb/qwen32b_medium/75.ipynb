````python
# Load the CSV file into a DataFrame
df = pd.read_csv('./input/microsoft.csv')

# Display the first few rows of the DataFrame to understand its structure
df.head()
````

[STEP GOAL]: Calculate the Daily Return column and add it to the DataFrame.


````python
# Calculate the Daily Return column
df['Daily Return'] = df['Close'].pct_change() * 100

# Display the DataFrame with the new 'Daily Return' column
df.head()
````

[STEP GOAL]: Compute the mean and standard deviation of the 'Daily Return' column.


````python
# Calculate mean and standard deviation of the 'Daily Return' column
mean_daily_return = df['Daily Return'].mean()
std_daily_return = df['Daily Return'].std()

# Round the results to two decimal places
mean_daily_return_rounded = round(mean_daily_return, 2)
std_daily_return_rounded = round(std_daily_return, 2)

# Display the mean and standard deviation
mean_daily_return_rounded, std_daily_return_rounded
````

The mean of the 'Daily Return' column is -0.14% and the standard deviation is 0.94%. This indicates that, on average, there is a slight daily decrease in the closing price of Microsoft stock, but with a relatively high volatility as measured by the standard deviation.
