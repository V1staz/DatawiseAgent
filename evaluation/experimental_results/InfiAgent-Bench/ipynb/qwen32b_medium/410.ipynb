````python
from scipy.stats import ks_2samp

# Perform the two-sample Kolmogorov-Smirnov test
ks_statistic, p_value = ks_2samp(male_non_survivors, female_non_survivors)
print(f'KS statistic: {ks_statistic}')
print(f'P-value: {p_value}')

# Interpret the result based on the p-value and alpha level of 0.05
alpha = 0.05
if p_value < alpha:
    print("The distributions are significantly different.")
else:
    print("The distributions are not significantly different.")
````
