[STEP GOAL]: Load the Credit.csv file into a Pandas DataFrame, then calculate the Pearson correlation coefficient between the "Limit" and "Balance" columns.


````python
# Load the dataset
credit_df = pd.read_csv('./input/Credit.csv')

# Display the first few rows of the DataFrame to understand its structure
credit_df.head()
````

````python
# Calculate the Pearson correlation coefficient between "Limit" and "Balance"
correlation_coefficient = credit_df['Limit'].corr(credit_df['Balance'], method='pearson')

# Round the result to two decimal places
correlation_coefficient_rounded = round(correlation_coefficient, 2)
correlation_coefficient_rounded
````

The Pearson correlation coefficient between the "Limit" and "Balance" columns in the Credit.csv dataset is 0.86, rounded to two decimal places.


The Pearson correlation coefficient between the "Limit" and "Balance" columns in the Credit.csv file has been calculated. The coefficient is 0.86, indicating a strong positive correlation between the limit on the credit card and the balance carried. This suggests that as the credit limit increases, the balance also tends to increase, which is consistent with the idea that individuals with higher credit limits may also carry higher balances.
