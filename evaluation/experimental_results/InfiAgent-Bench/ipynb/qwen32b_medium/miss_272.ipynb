## Prompt

```text
Question: Create a new feature named "TOTUSJZ_TOTUSJH_RATIO" by dividing the TOTUSJZ column by the TOTUSJH column. Calculate the mean and standard deviation of this new feature.
Add a small constant (1e-10) to the denominator (TOTUSJH column) to avoid dividing by zero.
```

## Expected Format

```text
Return 2 values: @mean_ratio[Mean of the TOTUSJZ_TOTUSJH_RATIO column, rounded to two decimal places], @stddev_ratio[Standard deviation of the TOTUSJZ_TOTUSJH_RATIO column, rounded to two decimal places].
```


[STEP GOAL]: Load the CSV file and create the new feature by dividing the TOTUSJZ column by the TOTUSJH column, adding a small constant to the denominator to avoid division by zero.


````python
# Import pandas library for data manipulation
import pandas as pd

# Load the dataset from the provided CSV file
data_path = './input/3901.csv'
df = pd.read_csv(data_path)

# Display the first few rows of the DataFrame to understand its structure
df.head()
````

[STEP GOAL]: Calculate the mean and standard deviation of the newly created feature.


````python
# Add a small constant to the denominator to avoid division by zero
small_constant = 1e-10

# Create the new feature by dividing TOTUSJZ by TOTUSJH + small_constant
df['TOTUSJZ_TOTUSJH_RATIO'] = df['TOTUSJZ'] / (df['TOTUSJH'] + small_constant)

# Calculate the mean and standard deviation of the new feature
mean_ratio = df['TOTUSJZ_TOTUSJH_RATIO'].mean()
std_dev_ratio = df['TOTUSJZ_TOTUSJH_RATIO'].std()

mean_ratio, std_dev_ratio
````

The new feature "TOTUSJZ_TOTUSJH_RATIO" has been successfully created by dividing the TOTUSJZ column by the TOTUSJH column (with a small constant added to the denominator to avoid division by zero). The mean of this new feature is approximately **22,756,785,531.29** and the standard deviation is approximately **969,133,356.79**.
