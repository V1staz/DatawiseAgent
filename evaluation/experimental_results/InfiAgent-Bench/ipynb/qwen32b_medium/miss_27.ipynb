[STEP GOAL]: Calculate the Interquartile Range (IQR) and identify outliers in the `charges` column using the 1.5xIQR rule.


````python
# Calculate the first quartile (Q1), third quartile (Q3), and IQR
Q1 = insurance_df['charges'].quantile(0.25)
Q3 = insurance_df['charges'].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = insurance_df[(insurance_df['charges'] < lower_bound) | (insurance_df['charges'] > upper_bound)]
outliers_count = len(outliers)

# Calculate mean and median of the outliers
mean_outliers = outliers['charges'].mean()
median_outliers = outliers['charges'].median()

outliers_count, mean_outliers, median_outliers
````

We have identified 139 outliers in the `charges` column with a mean charge of approximately $42,103.95 and a median charge of approximately $40,974.16.

This concludes the identification of outliers using the 1.5xIQR rule and the calculation of the mean and median charges of these outliers. We can now proceed to further analysis or visualization if required.

````python
from scipy import stats
import numpy as np

# Calculate Z-scores for the charges column
z_scores = stats.zscore(insurance_df['charges'])

# Identify outliers based on Z-scores
z_threshold = 3  # Typically, values outside ±3 standard deviations are considered outliers.
z_outliers = insurance_df[np.abs(z_scores) > z_threshold]

# Calculate the total number of outliers, and their mean and median charges
z_outliers_count = len(z_outliers)
z_mean_outliers = z_outliers['charges'].mean()
z_median_outliers = z_outliers['charges'].median()

z_outliers_count, z_mean_outliers, z_median_outliers
````
