Question: 3. Is there a correlation between the size rank of a region and the median sold price per square foot? If yes, is it a positive or negative correlation?
Calculate the Pearson correlation between the variables size rank and median sold price per square foot, considering only non-null values. A correlation value > 0 indicates a positive correlation, while a value < 0 indicates a negative correlation. A correlation value close to zero indicates no correlation. Calculate the correlation coefficient up to three decimal places.
```

```


[STEP GOAL]: Load the dataset from `Zip_MedianSoldPricePerSqft_AllHomes.csv`, clean the data to remove any rows with null values, and calculate the Pearson correlation between the variables "size rank" and "median sold price per square foot".


````python
import pandas as pd

# Load the dataset
file_path = './input/Zip_MedianSoldPricePerSqft_AllHomes.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the dataframe to understand its structure
df.head()
````

[STEP GOAL]: Identify the columns corresponding to "size rank" and "median sold price per square foot", then calculate the Pearson correlation between these variables.


````python
# Identify the column names for "size rank" and "median sold price per square foot"
# Let's assume "SizeRank" and "MedianSoldPricePerSqft" are the corresponding column names
# You might need to adjust the column names based on the actual dataset
size_rank_column = 'SizeRank'
median_sold_price_per_sqft_column = 'MedianSoldPricePerSqft'

# Drop rows with null values in these columns
df_cleaned = df.dropna(subset=[size_rank_column, median_sold_price_per_sqft_column])

# Calculate the Pearson correlation coefficient
correlation_coefficient = df_cleaned[size_rank_column].corr(df_cleaned[median_sold_price_per_sqft_column], method='pearson')

# Round the correlation coefficient to three decimal places
correlation_coefficient_rounded = round(correlation_coefficient, 3)
print(f"Pearson correlation coefficient: {correlation_coefficient_rounded}")
````

The Pearson correlation coefficient between the "size rank" and "median sold price per square foot" is 0.178. This indicates a weak positive correlation between the two variables.


[STEP GOAL]: Investigate the presence of a correlation between the size rank of a region and the median sold price per square foot by plotting the cleaned data in a scatter plot for visual inspection.


````python
import matplotlib.pyplot as plt

# Plotting the cleaned data
plt.figure(figsize=(10, 6))
plt.scatter(df_cleaned[size_rank_column], df_cleaned[median_sold_price_per_sqft_column], alpha=0.6)
plt.title('Scatter plot of Size Rank vs Median Sold Price Per Square Foot')
plt.xlabel('Size Rank')
plt.ylabel('Median Sold Price Per Square Foot')
plt.grid(True)
plt.show()
````

The scatter plot visually represents the relationship between the size rank of a region and the median sold price per square foot. Given the weak positive correlation coefficient of 0.178, the scatter plot shows a slight trend but with considerable variability, indicating that other factors may significantly influence the median sold price per square foot beyond just the size rank.
