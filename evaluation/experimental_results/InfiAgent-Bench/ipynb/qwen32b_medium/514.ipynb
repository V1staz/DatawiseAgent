[STEP GOAL]: Calculate the average review count for hotels in each city and compare these averages to the overall average. Identify cities where the average review count is more or less than twice the overall average.


````python
# Load the hotel_data.csv file into a DataFrame
hotel_df = pd.read_csv('./input/hotel_data.csv')

# Calculate the overall average review count of all hotels
overall_avg_review_count = hotel_df['review_count'].mean()

# Calculate the average review count for each city
average_review_count_per_city = hotel_df.groupby('city_name')['review_count'].mean().reset_index()

# Filter cities where the average review count is more or less than twice the overall average
threshold_high = 2 * overall_avg_review_count
threshold_low = 0.5 * overall_avg_review_count

high_review_cities = average_review_count_per_city[average_review_count_per_city['review_count'] > threshold_high]
low_review_cities = average_review_count_per_city[average_review_count_per_city['review_count'] < threshold_low]

print("Overall Average Review Count:", overall_avg_review_count)
print("\nAverage Review Count per City:")
print(average_review_count_per_city)

print("\nCities with Significantly Higher Review Counts:")
print(high_review_cities)

print("\nCities with Significantly Lower Review Counts:")
print(low_review_cities)
````

The calculations have been performed successfully. Here is a summary of the findings:

- The overall average review count of all hotels is approximately 1013.53.
- No cities have an average review count more than twice the overall average.
- Cities with significantly lower review counts (< half of the overall average) include:
  - Union City
  - Jersey City
  - Hoboken
  - Brooklyn

These cities have noticeably fewer reviews compared to the overall average.
