[STEP GOAL]: Load the Titanic dataset and filter it to include only passengers who survived and were in first class. Then calculate the Pearson correlation coefficient between the 'age' and 'fare' columns.


````python
import pandas as pd

# Load the Titanic dataset
file_path = './input/titanic.csv'
data = pd.read_csv(file_path)

# Filter the dataset to include only passengers who survived and were in first class
filtered_data = data[(data['Survived'] == 1) & (data['Pclass'] == 1)]

# Display the first few rows of the filtered data
filtered_data.head()
````

````python
# Calculate the Pearson correlation coefficient between 'Age' and 'Fare'
pearson_corr = filtered_data['Age'].corr(filtered_data['Fare'], method='pearson')
print(f"Pearson correlation coefficient between Age and Fare: {pearson_corr}")
````

The Pearson correlation coefficient between the 'Age' and 'Fare' columns for passengers who survived and were in first class is approximately `-0.123`. This suggests a weak negative linear relationship between age and fare for this group of passengers. 

It means that as age increases, there is a slight tendency for the fare to decrease, but the relationship is not very strong.
