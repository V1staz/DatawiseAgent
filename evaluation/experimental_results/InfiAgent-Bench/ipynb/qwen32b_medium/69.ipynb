## Prompt

```text
Question: Perform feature engineering by creating a new feature called "experience_score" that is calculated by multiplying the exper column with the looks column. Then, calculate the Pearson correlation coefficient between the "experience_score" feature and the wage column.
Create "experience_score" by directly multiplying values of exper and looks column. Calculate Pearson correlation coefficient between the new feature "experience_score" and wage. Correlation should be calculated up to three decimal places.
```

## Expected Format

```text
@correlation[correlation] where "correlation" is a number representing the correlation coefficient, rounded to three decimal places.
```


[STEP GOAL]: Load the dataset and create the new feature "experience_score".


````python
# Loading the dataset
df = pd.read_csv('./input/beauty and the labor market.csv')

# Creating the new feature "experience_score"
df['experience_score'] = df['exper'] * df['looks']

# Displaying the first few rows of the DataFrame to check the new feature
df.head()
````

[STEP GOAL]: Calculate the Pearson correlation coefficient between "experience_score" and the "wage" column.


````python
# Calculating the Pearson correlation coefficient
correlation = df['experience_score'].corr(df['wage'])

# Rounding the result to three decimal places
correlation_rounded = round(correlation, 3)

correlation_rounded
````

The Pearson correlation coefficient between the "experience_score" and the "wage" column is **0.252**.


The task has been successfully completed. We created a new feature named "experience_score" by multiplying the "exper" column with the "looks" column. Then, we calculated the Pearson correlation coefficient between the "experience_score" feature and the "wage" column, which resulted in a value of **0.252** (rounded to three decimal places).
