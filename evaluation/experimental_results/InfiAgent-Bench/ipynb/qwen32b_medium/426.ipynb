```text
Question: 2. What is the maximum sustained wind speed recorded during the storm with the highest maximum storm category?
Firstly, identify the storm with the highest maximum storm category, neglecting any ties. If there are multiple storms with the same highest maximum storm category, choose the one that appears first in the given dataset. Then find the maximum sustained wind speed corresponding to this particular storm.
```

## Expected Format

```text
@max_wind_speed[number] where "number" is a float with two decimal places indicating the highest wind speed recorded for the storm with the highest maximum storm category.
```


[STEP GOAL]: Load the file `cost_data_with_errors.csv` into a DataFrame and check its contents for initial understanding.


````python
import pandas as pd

# Load the dataset
file_path = './input/cost_data_with_errors.csv'
df = pd.read_csv(file_path)

# Display the first few rows of the dataframe
df.head()
````

The dataset has been successfully loaded and the first few rows have been displayed. We can see columns such as `name`, `dates_active`, `max_storm_cat`, `max_sust_wind`, among others. The column `max_storm_cat` seems to indicate the maximum storm category and `max_sust_wind` indicates the maximum sustained wind speed.

Now we'll move on to identifying the storm with the highest maximum storm category and finding the corresponding maximum sustained wind speed.


[STEP GOAL]: Identify the storm with the highest maximum storm category and retrieve the maximum sustained wind speed for this storm.


````python
# Find the storm with the highest storm category
highest_category_storm = df.loc[df['max_storm_cat'].idxmax()]

# Retrieve the maximum sustained wind speed for this storm
max_sustained_wind_speed = highest_category_storm['max_sust_wind']

max_sustained_wind_speed
````

The storm with the highest maximum storm category has a maximum sustained wind speed of 156.41568 knots. This value corresponds to the storm that had the highest storm category in the dataset.


[STEP GOAL]: Summarize the findings by reporting the storm name and its maximum sustained wind speed.


````python
# Storm name with the highest storm category
storm_name = highest_category_storm['name']
max_sustained_wind_speed

# Presenting the final result
result = f"The storm with the highest maximum storm category is '{storm_name}' and the maximum sustained wind speed recorded during this storm is {max_sustained_wind_speed:.2f} knots."
result
````

The storm with the highest maximum storm category is 'IRMA', and the maximum sustained wind speed recorded during this storm is 156.42 knots. This concludes our analysis as per the provided dataset.


The analysis has determined that the storm with the highest maximum storm category is named 'IRMA'. During this storm, the maximum sustained wind speed recorded was 156.42 knots. This fulfills the request to identify the storm with the highest maximum storm category and report the corresponding maximum sustained wind speed, based on the provided `cost_data_with_errors.csv` data.
