[STEP GOAL]: Load the auto-mpg.csv dataset from the input directory and calculate the Pearson correlation coefficient between the 'mpg' and 'weight' columns.


````python
# Load the auto-mpg.csv dataset
auto_mpg_path = "./input/auto-mpg.csv"
data = pd.read_csv(auto_mpg_path)

# Display the head of the DataFrame to understand its structure
data.head()
````

````python
# Calculate the Pearson correlation coefficient between 'mpg' and 'weight'
correlation_coefficient = data['mpg'].corr(data['weight'], method='pearson')
correlation_coefficient_rounded = round(correlation_coefficient, 2)
print(f"The Pearson correlation coefficient between 'mpg' and 'weight' is: {correlation_coefficient_rounded}")
````

The Pearson correlation coefficient between 'mpg' and 'weight' is -0.83, indicating a strong negative linear relationship between the two variables. This means that as the weight of the car increases, the miles per gallon (mpg) tends to decrease, which is an expected trend given the physical relationship between vehicle weight and fuel efficiency.


The Pearson correlation coefficient between the 'mpg' and 'weight' columns in the auto-mpg.csv dataset has been calculated and rounded to two decimal places. The result is -0.83, indicating a strong negative linear relationship between the two variables.
