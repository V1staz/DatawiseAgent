[STEP GOAL]: Extract the fares for each passenger class and conduct the Shapiro-Wilk test for normality.


````python
from scipy.stats import shapiro

# Define the significance level
alpha = 0.05

# Group the data by 'Pclass' and extract the 'Fare' column for each group
fare_by_class = [group['Fare'] for _, group in titanic_df.groupby('Pclass')]

# Conduct the Shapiro-Wilk Test for each passenger class
results = []
for i, fare in enumerate(fare_by_class):
    stat, p_value = shapiro(fare.dropna())  # drop NaN values before performing the test
    result = {
        'Passenger Class': i + 1,
        'Statistic': stat,
        'P-Value': p_value,
        'Normal Distribution?': p_value > alpha  # If P-Value > alpha, we do not reject the null hypothesis
    }
    results.append(result)

# Convert the results to a DataFrame
shapiro_results_df = pd.DataFrame(results)
print(shapiro_results_df)
````

The Shapiro-Wilk test results for the fare distribution among each passenger class are summarized below:

- **Passenger Class 1**: The p-value is extremely small (< 0.05), indicating that the fare distribution does not follow a normal distribution.
- **Passenger Class 2**: Similarly, the fare distribution for this class also does not follow a normal distribution due to a very low p-value.
- **Passenger Class 3**: Again, the fare distribution fails the test for normality with an exceptionally low p-value.

In summary, the fare distributions for all three passenger classes do not follow a normal distribution at the 0.05 significance level.
