[STEP GOAL]: Analyze the distribution of the 'Fare' column for each passenger class ('Pclass'). Calculate the mean, median, and population standard deviation of the fare for each class.


````python
# Group the dataframe by 'Pclass' and calculate the mean, median, and population standard deviation of the 'Fare' column for each group.
fare_stats = df.groupby('Pclass')['Fare'].agg(['mean', 'median', lambda x: x.std(ddof=0)]).round(2)

# Rename the columns for clarity
fare_stats.columns = ['Mean Fare', 'Median Fare', 'Population Std Dev']

# Display the resulting statistics
fare_stats
````

````python
# Correcting the issue where the Pclass column doesn't seem to contain the expected values 1, 2, and 3. This could be due to the Unnamed: 0 column causing confusion in the data frame structure. Let's remove this column and try again.
df = df.drop(columns=['Unnamed: 0'])

# Reattempt to analyze the distribution of the 'Fare' column for each passenger class ('Pclass').
fare_stats_corrected = df.groupby('Pclass')['Fare'].agg(['mean', 'median', lambda x: x.std(ddof=0)]).round(2)

# Rename the columns for clarity
fare_stats_corrected.columns = ['Mean Fare', 'Median Fare', 'Population Std Dev']

# Display the resulting statistics
fare_stats_corrected
````

````python
# Remove the row with Pclass value 0 since it seems to be an anomaly based on the context of the question.
df_cleaned = df[df['Pclass'] > 0]

# Recalculate the statistics with the cleaned DataFrame
fare_stats_cleaned = df_cleaned.groupby('Pclass')['Fare'].agg(['mean', 'median', lambda x: x.std(ddof=0)]).round(2)

# Rename the columns for clarity
fare_stats_cleaned.columns = ['Mean Fare', 'Median Fare', 'Population Std Dev']

# Display the corrected statistics
fare_stats_cleaned
````

[STEP GOAL]: Provide a visual representation of the fare distribution for each passenger class ('Pclass').


````python
import matplotlib.pyplot as plt

# Create a boxplot to visualize the fare distribution by passenger class
plt.figure(figsize=(10, 6))
boxplot = df_cleaned.boxplot(column='Fare', by='Pclass', grid=False)

# Set the title, labels, and remove the automatic title added by boxplot
plt.title('Fare Distribution by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.suptitle('')  # This removes the automatic title

# Show the plot
plt.show()
````
