````python
# Load the dataset
data_path = "./input/test_Y3wMUE5_7gLdaTN.csv"
df = pd.read_csv(data_path)

# Display the first few rows of the dataframe
df.head()
````

[STEP GOAL]: Calculate the Pearson correlation coefficient between ApplicantIncome and LoanAmount, ignoring rows with missing values for these columns, and round it to two decimal places.


````python
# Calculate the Pearson correlation coefficient
correlation_coefficient = df[['ApplicantIncome', 'LoanAmount']].corr(method='pearson').iloc[0, 1]

# Round the correlation coefficient to two decimal places
correlation_coefficient_rounded = round(correlation_coefficient, 2)

correlation_coefficient_rounded
````
