## Prompt

```text
Question: Create a new feature in the Credit.csv file by calculating the ratio of "Balance" to "Limit" for each individual.
Calculate the ratio as Balance / Limit. For any individual with a Limit of zero, their ratio should be defined as zero to avoid division by zero.
```

## Expected Format

```text
@addedfeature[ratio], where "ratio" refers to the newly created column containing the ratio of balance to limit for each individual, with a precision of two decimal places for each individual's ratio data.
```


[STEP GOAL]: Load the Credit.csv file into a DataFrame and calculate the new feature which is the ratio of "Balance" to "Limit". Set the ratio to 0 for individuals with a Limit of zero to avoid division by zero.


````python
import pandas as pd

# Load the dataset
credit_df = pd.read_csv('./input/Credit.csv')

# Display the first few rows of the dataset
credit_df.head()
````

````python
# Calculate the new feature: Balance to Limit ratio
# Avoid division by zero by setting ratio to 0 for rows with Limit == 0
credit_df['Balance_to_Limit_Ratio'] = credit_df.apply(lambda row: row['Balance'] / row['Limit'] if row['Limit'] != 0 else 0, axis=1)

# Display the updated DataFrame with the new feature
credit_df[['Balance', 'Limit', 'Balance_to_Limit_Ratio']].head()
````

The new feature "Balance_to_Limit_Ratio" has been successfully added to the DataFrame. This feature represents the ratio of the "Balance" column to the "Limit" column for each individual. For individuals with a Limit of zero, the ratio has been set to 0 to prevent division by zero.


[STEP GOAL]: Save the updated DataFrame back to a CSV file so that the newly created feature "Balance_to_Limit_Ratio" is included in the output file.


````python
# Save the updated DataFrame back to a CSV file
output_file_path = './working/Credit_Updated.csv'
credit_df.to_csv(output_file_path, index=False)

# Confirm that the file has been saved correctly
!ls -l $output_file_path
````
