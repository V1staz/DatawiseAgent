````python
# Load the weather data from the CSV file into a DataFrame
df_weather = pd.read_csv('./input/weather_data_1864.csv')

# Display the first few rows of the DataFrame to ensure the correct loading
df_weather.head()
````

````python
# Calculate the IQR for the 'obs_value' column
Q1 = df_weather['obs_value'].quantile(0.25)
Q3 = df_weather['obs_value'].quantile(0.75)
IQR = Q3 - Q1

print(f"Q1: {Q1}, Q3: {Q3}, IQR: {IQR}")
````

````python
# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify and count the outliers
outliers = df_weather[(df_weather['obs_value'] < lower_bound) | (df_weather['obs_value'] > upper_bound)]
outlier_count = len(outliers)

print(f"The number of outliers in the 'obs_value' column is: {outlier_count}")
````
